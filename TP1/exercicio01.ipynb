# 1. List comprehension

```python
# Retorna os caracteres individuais de uma string que não são caracteres de espaço em branco
def caracteresString(texto):
    # "for c" -> variável de iteração
    # "in texto" -> string sendo percorrida
    # "if not c.isspace()]" ->  condição
    caracteres = [c for c in texto if not c.isspace()]
    return caracteres


texto = "Sítio do pica-pau amarelo \n 2023"
result = caracteresString(texto)
print(result)
```
