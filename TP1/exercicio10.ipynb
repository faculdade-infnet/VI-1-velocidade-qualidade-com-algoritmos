### 10. Bubble Sort - Ordenação Alfabética

```python
def bubble_sort_alfabetico(lista):
  n = len(lista)

  # Loop exteno, controla o número de passagens do algoritmo.
  for i in range(n):
    trocou = False
    # Loop interno para controlar as comparaçoes entre elementos.
    # Último índice válido para comparação
    # [n(numero de elementos na lista) - i(contador atual) - 1(lista inicia em 0)]
    for j in range(0, n -i -1):   # Start index, Stop index
      # Se estiver fora de ordem → troca.
      if lista[j].lower() > lista[j + 1].lower():
        # Troca os elementos
        lista[j], lista[j + 1] = lista[j + 1], lista[j]
        trocou = True
    if not trocou:
      break


# testando
lista1 = ["az", "lima", "xpto9", "rio", "code"]
bubble_sort_alfabetico(lista1)
print(f'lista 1 = {lista1}')

lista2 = ["sol", "br7", "alpha", "zen", "byte"]
bubble_sort_alfabetico(lista2)
print(f'lista 2 = {lista2}')

lista3 = ["lua", "delta9", "sky", "Matrix", "go", "nova", "fluxo", "macerati", "zenith"]
bubble_sort_alfabetico(lista3)
print(f'lista 3 = {lista3}')

lista4 = ["ai", "tiger", "Bravo", "bastardo", "io", "quantum", "sky7", "atlas", "neo"]
bubble_sort_alfabetico(lista4)
print(f'lista 4 = {lista4}')

lista5 = ["pi", "gamma", "rocket", "bit", "vector9", "code", "Boca", "omega", "data", "flux", "ai7", "storm"]
bubble_sort_alfabetico(lista5)
print(f'lista 5 = {lista5}')
```
