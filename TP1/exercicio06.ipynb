# 6 - Tabuleiro

```python
# Calcula em qual quadrado do tabuleiro você precisará colocar uma determinada quantidade de grãos de arroz informada
def tabuleiro(numeroGraos):
    quadrados = 1

    while numeroGraos > 1:
      numeroGraos = numeroGraos /2
      quadrados = quadrados + 1
    return quadrados


posicaoQuadrado = tabuleiro(32)
print(posicaoQuadrado)
```
