### 9. Bubble Sort - Ordenação Numérica

```python
import random

def bubble_sort_numerico(lista):
  n = len(lista)

  # Loop exteno, controla o número de passagend do algoritmo.
  for i in range(n):
    trocou = False
    # Loop interno para controlar as comparaçoes entre elementos.
    # Último índice válido para comparação
    # [n(numero de elementos na lista) - i(contador atual) - 1(lista inicia em 0)]
    for j in range(0, n -i -1):   # Start index, Stop index
      # Se estiver fora de ordem → troca.
      if lista[j] > lista[j + 1]:
        # Troca os elementos
        lista[j], lista[j + 1] = lista[j + 1], lista[j]
        trocou = True
    if not trocou:
      break


# testando
listas = [
    [random.randint(0, 100) for _ in range(10)],
    [random.randint(0, 100) for _ in range(10)],
    [random.randint(0, 100) for _ in range(15)],
    [random.randint(0, 100) for _ in range(15)],
    [random.randint(0, 100) for _ in range(20)]
]

for i, lista in enumerate(listas, start=1):
    bubble_sort_numerico(lista)
    print(f'lista {i} = {lista}')
```
