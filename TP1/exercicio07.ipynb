# 7. Notação Big O

```python
# Cria lista com strings que iniciam com a letra "a"
def selectAStrings (array):
    newArray = []

    for i in range(len(array)):
        if array[i].startswith("a"):
            newArray.append(array[i])
    return newArray


# testando
lista = ["lua", "delta9", "sky", "Matrix", "go", "nova", "ameixa", "macerati", "abacate"]
resultado = selectAStrings(lista)
print(resultado)
```
