# 5 - Rescrever função de O(n²)  -> O(n)

### 5.1 - Função O(n²)

```python
# Encontrar o maior número único dentro de um array
def greatestNumber(array):
    for i in array:
        # Suponha, por enquanto, que i seja o maior:
        isMaiorValor = True

        for j in array:
            # Se encontrarmos outro valor maior que i, então, i não é o maior:
            if j > i:
                MaiorValor = i
                isMaiorValor = False

        # Se, depois de verificarmos todos os outros números, i
        # i ainda for o maior, significa que i é o maior número:
        if isMaiorValor:
            return i


# testando
lista = [3, 15, 89, 15, 62, 38, 91, 91, 10, 55]
maiorValor = greatestNumber(lista)
print(f'O maior valor é: {maiorValor}')
```

### 5.2 - Função O(n)

```python
# Encontrar o maior número único dentro de um array
def greatestNumber(array):
    maiorValor = array[0]

    for j in range(len(array) - 1):
        # Compara o maiorValor atual com o valor seguinte
        if array[j+1] > maiorValor:
            # print(array[j+1])
            maiorValor = array[j+1]

    return maiorValor


# testando
lista = [3, 15, 89, 15, 62, 38, 91, 91, 10, 55]
maiorValor = greatestNumber(lista)
print(f'O maior valor é: {maiorValor}')
```
