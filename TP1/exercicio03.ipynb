# 3. Número de passos para encontrar número 8 com Busca Linear

```python
def busca_linear(lista, busca):
    n = len(lista)
    cont =  0 # conta número de passos executados

    for i in range(n):
        cont += 1
        if lista[i] == busca:
            print("Número de passos:", cont)
            return



# testando
lista = [2, 4, 6, 8, 10, 12, 13]
busca_linear(lista, 8)
print(lista)
```
