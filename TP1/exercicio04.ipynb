# 4. Número de passos para encontrar número 8 com Busca Binária

```python
def busca_binaria(lista, busca):
    inicio = 0
    fim = len(lista)-1
    cont =  0 # conta número de passos executados

    while inicio <= fim:
        cont += 1
        meio = (inicio + fim)//2   # divisão inteira

        if lista[meio] == busca:
            print("Número de passos:", cont)
            print("Número encontrado: ", lista[meio])
            return
        if lista[meio] > busca:
            fim = meio - 1
        else:
            inicio = meio + 1
    return -1



# testando
lista = [2, 4, 6, 8, 10, 12, 13]
busca_binaria(lista, 8)
print(lista)
```
