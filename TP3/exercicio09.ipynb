# 9 - Implementação do QuickSort


```python
# 1 - Escolhe um elemento da lista como pivô
# 2 - Divide a lista em 2 o 3 partes:
#     menores ou iguais ao pivô e maiores que o pivô
# 3 - Ordena recursivamente as partes menores e maiores
# 4 - Junta: menores ordenados + pivôs + maiores ordenados
import time
import random
def quick_sort(lista):
    # CASO BASE: se a lista tem 0 ou 1 elemento, ela já está ordenada
    if len(lista) <= 1:
        return lista

    pivo = lista[-1]     # Escolhe o último elemento como pivô

    # Cria listas menores com list comprehension
    menores = [x for x in lista[:-1] if x <= pivo]

    # Cria listas maiores com list comprehension
    maiores = [x for x in lista[:-1] if x > pivo]

    # dividir + resolver recursivamente + combinar
    return quick_sort(menores) + [pivo] + quick_sort(maiores)


# TESTANDO
listas = [
    ("lista ordenada", list(range(1, 21, 1))),
    ("lista ordenada", list(range(20, 41, 1))),
    ("lista invertida", list(range(60, 0, -2))),
    ("lista invertida", list(range(80, 0, -3))),
    ("lista aleatória", random.sample(range(1, 100), 20)),
    ("lista aleatória", random.sample(range(1, 100), 20))
]

resultados = []

for i, (nome, lista) in enumerate(listas, start=1):
    print(f"{i} - {nome}:")

    inicio_t = time.perf_counter()
    lista_ordenada = quick_sort(lista)
    tempo = time.perf_counter() - inicio_t

    print(f"\tTempo: {tempo:.8f} segundos = {lista_ordenada}")
    resultados.append((nome, tempo))
# =========================
# ANÁLISE FINAL
# =========================

print("\nANÁLISE DE DESEMPENHO\n")

# Ordena do mais rápido para o mais lento
resultados.sort(key=lambda x: x[1])

melhor_tempo = resultados[0][1]

for i, (nome, tempo) in enumerate(resultados, start=1):
    diferenca = ((tempo - melhor_tempo) / melhor_tempo) * 100

    if i == 1:
        print(f"{i}º MAIS RÁPIDO → {nome} | tempo = {tempo:.8f} s | referência (0%)")
    else:
        print(f"{i}º → {nome} | tempo = {tempo:.8f} s | {diferenca:.2f}% mais lento")
```
