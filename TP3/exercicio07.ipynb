# 7 - Análise de Complexidade em Algoritmos Recursivos

```python
def produto(n1, n2):
    # verifica se algum número é negativo
    if n1 < 0 or n2 < 0:
        return None

    # CASO BASE: quando n2 chega a 0 a recursão termina
    if n2 == 0:
        return 0

    # soma n1 com o resultado da próxima chamada recursiva
    resultado = n1 + produto(n1, n2 - 1)
    return resultado   # resultado da multiplicação


# TESTANDO
n1 = 3      # primeiro número da multiplicação
n2 = 3      # segundo número da multiplicação
resultado = produto(n1, n2)
print(f'\nA multiplição por soma de {n1} x {n2} = {resultado}\n')

print("""A questão escolhida foi a questão 1 (Multiplicação Recursiva por Soma).

ANÁLISES:
Chamada redundante: Não existem chamadas redundantes; cada chamada recursiva é necessária para somar n1, n2 vezes.
Crescimento do número de chamadas: A função é chamada recursivamente uma vez para cada decremento de n2 até atingir 0. Se n2 = 5, haverá 5 chamadas. O número de chamadas cresce linearmente com n2.
Complexidade (Big O): O(n2), pois o número de chamadas e operações é proporcional a n2, ou seja, a complexidade cresce segundo o valor de n2.""")
```
