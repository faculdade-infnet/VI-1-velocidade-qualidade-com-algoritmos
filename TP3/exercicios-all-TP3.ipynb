# 1 - Multiplicação Recursiva por Soma

```python
def produto(n1, n2):
    print(f"Chamada produto({n1} x {n2})")

    # verifica se algum número é negativo
    if n1 < 0 or n2 < 0:
        print(f"Não é permitido números negativos")
        return None

    # CASO BASE: quando n2 chega a 0 a recursão termina
    if n2 == 0:
        print("Return 0")
        return 0

    # soma n1 com o resultado da próxima chamada recursiva
    resultado = n1 + produto(n1, n2 - 1)
    print(f"Retorno {resultado} para produto({n1} x {n2})")
    return resultado   # resultado da multiplicação


# TESTANDO
n1 = 2      # primeiro número da multiplicação
n2 = 3      # segundo número da multiplicação
resultado = produto(n1, n2)
print(f'\nA multiplição por soma de {n1} x {n2} = {resultado}')
# observação: a soma ocorre no retorno
```

# 2 - Fatoração Recursiva de Inteiros

```python
import math

# x = número que queremos fatorar
# lowest = menor divisor que vamos testar
def factor(x, lowest=2):

    # CASOS ESPECIAIS
    if x == 0 or x ==1 or x < 0:
        return "Casos especiais"

    # limite de busca até a raiz quadrada
    limite = int(math.sqrt(x))

    # testa divisores entre lowest e √x
    for i in range(lowest, limite + 1):
        # se encontrou divisor com resto = 0
        if x % i == 0:
            # adiciona o fator e continua fatorando o resto
            # "x // i" dividie o número pelo fator encontrado → será o novo x
            # ", i)" passa a ser o novo lowest (menor divisor a testar)
            # "+" junta as duas istas de fatores
            return [i] + factor(x // i, i)

    # se nenhum divisor foi encontrado → número primo
    return [x]


# TESTANDO
testes = [72, 64, 102, 125, 18, 36, 24]

for n in testes:
    print(f"Fatores de {n} -> {factor(n)}")
```

# 3 -  Potenciação Recursiva com Expoentes Negativos

```python
# IDEIA CENTRAL: calcular potência usando recursão
# 1 - verificar casos base para parar a recursão
# 2 - tratar expoente negativo
# 3 - diminuir o expoente a cada chamada recursiva
# 4 - multiplicar a base até o expoente chegar ao caso base
# x = base da potência
# y = expoente da potência
def power(x, y):
    if x == 0 and y < 0:        # CASO BASE: evita divisão por zero quando base 0 tem expoente negativo
        return "Indefinido"

    if y == 0:                  # CASO BASE: qualquer número elevado a 0 é 1
        return 1

    if y == 1:                  # CASO BASE: potência de expoente 1 retorna a própria base
        return x

    # se expoente é negativo
    if y < 0:
        return 1 / power(x, -y)  # transforma o expoente -y para y e calcula o inverso
    else:
        return x * power(x, y - 1 )  # "power()" chamaa função com um expoente menor

# TESTANDO
lista_bases = [0, 1, 2, 5, 8]         # listas bases
lsita_potencia = [-2, 0, 1, 2, -3]    # lsitas expoentes

# percorre todas as bases
for i in lista_bases:
    for potencia in lsita_potencia:     # percorre todos os expoentes
        print(f"Potência de {i}^{potencia} = {power(i, potencia)}")

# Complexidade O(|y|) porque a função chama ela mesma |y| vezes
```

# 4 - Navegação Recursiva em Estrutura de Diretórios

```python
# 1 - listar itens da pasta
# 2 - para cada item:
#      se for arquivo → mostrar
#      se for pasta → mostrar e entrar nela (recursão)
import os

def navegar(diretorio, nivel=0):
    # lista todos os itens(diretótios/pastas) dentro do diretório
    itens = os.listdir(diretorio)

    # percorre todos os itens(diretótios/pastas) dentro
    for item in itens:
        # nome da pasta(diretorio = "docs") + nome do arquivo(item = "arquivo.txt") ou subpasta = caminho completo
        caminho = os.path.join(diretorio, item)
        print("  " * nivel + item)      # indentação conforme o nível

        # se for pasta, entra nela
        if os.path.isdir(caminho):
            navegar(caminho, nivel + 1)  # (recursão)

# TESTANDO -----------------------------------------------------------------
navegar("/content")
```

# 5 - Geração Recursiva de Combinações (Times)

```python
# IDEIA CENTRAL: Você tem uma lista de pessoas (candidates) e quer formar times de tamanho k.
# 1 - percorrer os candidatos disponíveis
# 2 - escolher um candidato para entrar no time
# 3 - chamar a função novamente para completar o time (recursão)
# 4 - quando o time tiver k pessoas → mostrar a combinação
# 5 - continuar testando outras combinações sem repetir candidatos
# candidates = lista de nomes
# k = nº máximo de intens na lista,
# inicioo = é o ponto de onde o índice vai começar no loop
# atual = nomes que ja foram escolhidos
def teams(candidates, k, inicio=0, atual=""):
    # verifica se o time atual já tem k(quantidade de pessoas p/ completar o time)
    # len(atual.split()): separa o time atual ple ospaço entre as palavras e entao conta quantas palavras tem,
    if len(atual.split()) == k:
        print(atual.strip())    # mostra o time formado
        return

    # percorre os candidatos disponíveis na lista
    for i in range(inicio, len(candidates)):
        # recursividasde, adicionao o candidato atual ao time
        # i + 1 garante que não vamos repetir candidatos anteriores
        # candidates, k, i + 1 = inicio, atual = lista atual já formada
        teams(candidates, k, i + 1, atual + " " + candidates[i])


# TESTANDO--------------------------------------------------------
candidates = ["Ana", "Bruno", "Carlos", "Diana", "Ricado"]
k = 2
teams(candidates, k)
```

# 6 - Problema da Mochila (Knapsack Recursivo)

```python
# IDEIA CENTRAL: encontrar combinações de pesos que somem exatamente ao valor alvo (target)
# 1 - olhar o peso atual da lista
# 2 - decidir se vamos usar ou não esse peso
# 3 - chamar a função novamente para continuar testando (recursão)
# 4 - quando a soma chegar exatamente ao target → guardar a combinação
# 5 - quando passar do target ou acabar a lista → parar aquela tentativa

# target = valor a ser atingido
# weights = lista de pesos
# index = posição atual
# atual = pesos já escolhidos na combinação
# resultados = lista com combinações válidas
def knapsack(target, weights, index=0, atual=None, resultados=None):
    if atual is None:       # cria lista vazia para armazenar pesos escolhidos
        atual = []

    if resultados is None:  # cria lista para armazenar todas as combinações válidas
        resultados = []

    # CASO BASE: se o target chegar a 0, encontramos uma combinação válida
    if target == 0:
        resultados.append(atual.copy())  # guarda a combinação encontrada
        return resultados

    # CASO BASE: se o target ficar negativo ou acabar os pesos
    if target < 0 or index >= len(weights):
        return resultados  # encerra essa tentativa

    # 1: adiciona peso atual na combinação
    # target: weights[index]: diminui o target pelo peso escolhido ,
    # weights: lsita de pessos
    # index + 1: cham o proximo índice pro peso
    #  atual + [weights[index]]: adiciona o peso atual na lista
    knapsack(target - weights[index], weights, index + 1, atual + [weights[index]], resultados)

    # 2: NÃO adiciona peso atual na combinação
    knapsack(target, weights, index + 1, atual, resultados)

    return resultados   # retorna todas as combinações encontradas


# TESTANDO -------------------------------------------
weights = [1, 2, 3, 4, 5, 7]    # pesos disponíveis
target = 10                     # valor que queremos atingir

resultado = knapsack(target, weights)
print("Combinações que somam", target)  # combinações encontradas

# precorre todos os resultados
if len(resultado) == 0:
    print("Nenhuma combinação encontrada")
else:
    for i in resultado:
        print(i)
```

# 7 - Análise de Complexidade em Algoritmos Recursivos

```python
def produto(n1, n2):
    # verifica se algum número é negativo
    if n1 < 0 or n2 < 0:
        return None

    # CASO BASE: quando n2 chega a 0 a recursão termina
    if n2 == 0:
        return 0

    # soma n1 com o resultado da próxima chamada recursiva
    resultado = n1 + produto(n1, n2 - 1)
    return resultado   # resultado da multiplicação


# TESTANDO
n1 = 3      # primeiro número da multiplicação
n2 = 3      # segundo número da multiplicação
resultado = produto(n1, n2)
print(f'\nA multiplição por soma de {n1} x {n2} = {resultado}\n')

print("""A questão escolhida foi a questão 1 (Multiplicação Recursiva por Soma).

ANÁLISES:
Chamada redundante: Não existem chamadas redundantes; cada chamada recursiva é necessária para somar n1, n2 vezes.
Crescimento do número de chamadas: A função é chamada recursivamente uma vez para cada decremento de n2 até atingir 0. Se n2 = 5, haverá 5 chamadas. O número de chamadas cresce linearmente com n2.
Complexidade (Big O): O(n2), pois o número de chamadas e operações é proporcional a n2, ou seja, a complexidade cresce segundo o valor de n2.""")
```

# 8 - Otimização com Programação Dinâmica

### 8.1. - Versão recursiva

```python
# Recusividade Direta - "n" representa a posição na sequência de Fibonacci que queremos calcular
def fibonacci_rec(n, chamadas=None):
    if chamadas is None:
        chamadas = {"total": 0}  # contador de chamadas

    chamadas["total"] += 1  # registra cada chamada da função

    # CASOS BASE da recursividade, na sequência de Fibonacci: F(1) e F(2) = 1
    if n == 1 or n == 2:
        return 1, chamadas["total"]  # retorna valor e total de chamadas
    else:
        # valor atual é a soma dos dois anteriores
        # função chama ela mesma duas vezes, p/ calcular F(n-1) e p/ calcular F(n-2)
        a, _ = fibonacci_rec(n-1, chamadas)
        b, _ = fibonacci_rec(n-2, chamadas)

        return a + b, chamadas["total"]
```

### 8.2. - Programação Dinâmica

```python
# Fibonacci com Programação Dinâmica (Memoização)
def fibonacci_pd(n, memo=None, chamadas=None):
    if memo is None:
        memo = {}  # dicionário para armazenar resultados já calculados
    if chamadas is None:
        chamadas = {"total": 0}  # contador de chamadas

    chamadas["total"] += 1  # registra cada chamada

    # se o valor já foi calculado, retorna do memo
    if n in memo:
        return memo[n], chamadas["total"]

    # casos base
    if n == 1 or n == 2:
        memo[n] = 1
        return 1, chamadas["total"]

    # chama recursivamente com memo
    a, _ = fibonacci_pd(n-1, memo, chamadas)
    b, _ = fibonacci_pd(n-2, memo, chamadas)

    memo[n] = a + b
    return memo[n], chamadas["total"]
```

### 8.3 - Test

```python
# TESTE COMPARATIVO -----------------------------------------------------
lista = [3, 5, 6, 8, 10]
print("COMPARAÇÃO FIBONACCI:")
print(f"{'':<16}Recursivo{'':<4}vs {'':<4}Programação Dinâmica")

for item in lista:
    # Recursão Simples
    for i in range(1, item + 1):
        resultado_rec, total_chamadas_rec = fibonacci_rec(i)

    # Com Memoização
    for i in range(1, item + 1):
        resultado_pd, total_chamadas_pd = fibonacci_pd(i)

    print(f"{f'Fib({item}) = {resultado_rec}':<15} chamadas {total_chamadas_rec:<3} | {total_chamadas_pd} chamadas")
```

# 9 - Implementação do QuickSort


```python
# 1 - Escolhe um elemento da lista como pivô
# 2 - Divide a lista em 2 o 3 partes:
#     menores ou iguais ao pivô e maiores que o pivô
# 3 - Ordena recursivamente as partes menores e maiores
# 4 - Junta: menores ordenados + pivôs + maiores ordenados
import time
import random
def quick_sort(lista):
    # CASO BASE: se a lista tem 0 ou 1 elemento, ela já está ordenada
    if len(lista) <= 1:
        return lista

    pivo = lista[-1]     # Escolhe o último elemento como pivô

    # Cria listas menores com list comprehension
    menores = [x for x in lista[:-1] if x <= pivo]

    # Cria listas maiores com list comprehension
    maiores = [x for x in lista[:-1] if x > pivo]

    # dividir + resolver recursivamente + combinar
    return quick_sort(menores) + [pivo] + quick_sort(maiores)


# TESTANDO
listas = [
    ("lista ordenada", list(range(1, 21, 1))),
    ("lista ordenada", list(range(20, 41, 1))),
    ("lista invertida", list(range(60, 0, -2))),
    ("lista invertida", list(range(80, 0, -3))),
    ("lista aleatória", random.sample(range(1, 100), 20)),
    ("lista aleatória", random.sample(range(1, 100), 20))
]

resultados = []

for i, (nome, lista) in enumerate(listas, start=1):
    print(f"{i} - {nome}:")

    inicio_t = time.perf_counter()
    lista_ordenada = quick_sort(lista)
    tempo = time.perf_counter() - inicio_t

    print(f"\tTempo: {tempo:.8f} segundos = {lista_ordenada}")
    resultados.append((nome, tempo))
# =========================
# ANÁLISE FINAL
# =========================

print("\nANÁLISE DE DESEMPENHO\n")

# Ordena do mais rápido para o mais lento
resultados.sort(key=lambda x: x[1])

melhor_tempo = resultados[0][1]

for i, (nome, tempo) in enumerate(resultados, start=1):
    diferenca = ((tempo - melhor_tempo) / melhor_tempo) * 100

    if i == 1:
        print(f"{i}º MAIS RÁPIDO → {nome} | tempo = {tempo:.8f} s | referência (0%)")
    else:
        print(f"{i}º → {nome} | tempo = {tempo:.8f} s | {diferenca:.2f}% mais lento")
```

# 10 - Instrumentação do QuickSort


```python
# 1 - Escolhe um elemento da lista como pivô
# 2 - Divide a lista em 2 o 3 partes:
#     menores ou iguais ao pivô e maiores que o pivô
# 3 - Ordena recursivamente as partes menores e maiores
# 4 - Junta: menores ordenados + pivôs + maiores ordenados
import time
import random

comparacoes = 0
copias = 0
def quick_sort(lista):
    global comparacoes, copias      # impede que as variáveis sejam recriadas localmente a durante a recursão
    menores = []
    maiores = []

    # CASO BASE: se a lista tem 0 ou 1 elemento, ela já está ordenada
    if len(lista) <= 1:
        return lista

    pivo = lista[len(lista)-1]     # escolhe o último elemento como pivô

    for x in lista[:-1]:
        comparacoes += 1
        # se testamos algo é uma comparação
        if x <= pivo:
            menores.append(x)
            copias += 1  # se adicionamos um elemento na lista estamos fazendo cópias
        else:
            maiores.append(x)
            copias += 1  # se adicionamos um elemento na lista estamos fazendo cópias

    # juntar listas, também é cópia
    copias += len(menores) + len(maiores) + 1

    # dividir + resolver recursivamente + combinar
    return quick_sort(menores) + [pivo] + quick_sort(maiores)


# TESTANDO
listas = [
    ("lista ordenada", list(range(1, 21, 1))),
    ("lista ordenada", list(range(20, 41, 1))),
    ("lista invertida", list(range(60, 0, -2))),
    ("lista invertida", list(range(80, 0, -3))),
    ("lista aleatória", random.sample(range(1, 100), 20)),
    ("lista aleatória", random.sample(range(1, 100), 20))
]

resultados = []

for i, (nome, lista) in enumerate(listas, start=1):
    comparacoes = 0
    copias = 0

    print(f"{i} - {nome}:")

    inicio = time.perf_counter()
    quick_sort(lista)
    tempo = time.perf_counter() - inicio

    print(f"\tTempo: {tempo:.8f} s")
    print(f"\tComparações: {comparacoes}")
    print(f"\tCópias: {copias}")

    resultados.append((nome, tempo, comparacoes, copias))
```

# 11 - Implementação do QuickSelect

```python
# 1 - Escolhe um elemento da lista como pivô
# 2 - Particiona a lista em menores ou iguais ao pivô e maiores que o pivô
# 3 - Coloca o pivô na posição correta que teria se a lista estivesse ordenada
# 4 - Verifica se o pivô é o k-ésimo elemento procurado
# 5 - Continua recursivamente apenas no lado da lista onde o k-ésimo elemento pode estar
def particao(lista, esq, dir):
    pivo = lista[dir]  # escolhe o último elemento da lista como pivô
    i = esq - 1        # índice que indica a posição final dos elementos menores que o pivô

    # percorre a lista do índice 'esq' até 'dir'
    for j in range(esq, dir):
        # se o elemento atual for menor ou igual ao pivô, menores a esquerda
        if lista[j] <= pivo:
            i += 1  # avança o índice dos menores
            # troca o elemento atual com o elemento na posição i
            # isso move o valor menor que o pivô para a parte esquerda da lista
            lista[i], lista[j] = lista[j], lista[i]

    # após percorrer todos os elementos, coloca o pivô na posição correta
    # todos os elementos menores ficam à esquerda e os maiores à direita
    lista[i + 1], lista[dir] = lista[dir], lista[i + 1]

    # retorna o índice final onde o pivô foi colocado
    return i + 1


# Função principal do QuickSelect que encontra o k-ésimo menor elemento
def quickselect(lista, esq, dir, k):
    if esq <= dir:      # verifica se ainda existe uma parte válida da lista para analisar

        # chama a função de partição para reorganizar a lista e obter a posição final do pivô
        pivo_index = particao(lista, esq, dir)

        # sinterrompe ao encontrae o k-ésimo menor elemento
        if pivo_index == k:
            return lista[pivo_index]

        # se a posição do pivô for maior que k, então o elemento procurado está na parte esquerda da lista
        elif pivo_index > k:
            return quickselect(lista, esq, pivo_index - 1, k)
        else:                        # caso contrário, o elemento está na parte direita da lista
            return quickselect(lista, pivo_index + 1, dir, k)

    return None         # caso a busca não encontre o elemento (situação limite)


# TESTANDO ------------------------------------------------------------------------
teste = [6, 4, 7, 1, 3, 8, 5, 2]
mediana = len(teste) // 2
resultado = quickselect(teste, 0, len(teste) - 1, mediana)

print("Mediana:", resultado)
```

# 12 - Comparação entre QuickSort e QuickSelect

### 12.1. - QuickSort

```python
# 1 - Escolhe um elemento da lista como pivô
# 2 - Divide a lista em 2 o 3 partes:
#     menores ou iguais ao pivô e maiores que o pivô
# 3 - Ordena recursivamente as partes menores e maiores
# 4 - Junta: menores ordenados + pivôs + maiores ordenados
import time
import random

comparacoes_qselect = 0
copias = 0
def quick_sort(lista):
    global comparacoes_qselect, copias      # impede que as variáveis sejam recriadas localmente a durante a recursão
    menores = []
    maiores = []

    # CASO BASE: se a lista tem 0 ou 1 elemento, ela já está ordenada
    if len(lista) <= 1:
        return lista

    pivo = lista[len(lista)-1]     # escolhe o último elemento como pivô

    for x in lista[:-1]:
        comparacoes_qselect += 1
        # se testamos algo é uma comparação
        if x <= pivo:
            menores.append(x)
            copias += 1  # se adicionamos um elemento na lista estamos fazendo cópias
        else:
            maiores.append(x)
            copias += 1  # se adicionamos um elemento na lista estamos fazendo cópias

    # juntar listas, também é cópia
    copias += len(menores) + len(maiores) + 1

    # dividir + resolver recursivamente + combinar
    return quick_sort(menores) + [pivo] + quick_sort(maiores)
```

### 12.2. Quick Select

```python
# 1 - Escolhe um elemento da lista como pivô
# 2 - Particiona a lista em menores ou iguais ao pivô e maiores que o pivô
# 3 - Coloca o pivô na posição correta que teria se a lista estivesse ordenada
# 4 - Verifica se o pivô é o k-ésimo elemento procurado
# 5 - Continua recursivamente apenas no lado da lista onde o k-ésimo elemento pode estar
comparacoes_qselect = 0
copias = 0

def particao(lista, esq, dir):
    global comparacoes_qselect, copias

    pivo = lista[dir]  # escolhe o último elemento da lista como pivô
    i = esq - 1        # índice que indica a posição final dos elementos menores que o pivô

    # percorre a lista do índice 'esq' até 'dir'
    for j in range(esq, dir):
        comparacoes_qselect += 1
        # se o elemento atual for menor ou igual ao pivô
        if lista[j] <= pivo:
            i += 1  # avança o índice dos menores
            # troca o elemento atual com o elemento na posição i
            # isso move o valor menor que o pivô para a parte esquerda da lista
            lista[i], lista[j] = lista[j], lista[i]
            copias += 1

    # após percorrer todos os elementos, coloca o pivô na posição correta
    # todos os elementos menores ficam à esquerda e os maiores à direita
    lista[i + 1], lista[dir] = lista[dir], lista[i + 1]
    copias += 1

    # retorna o índice final onde o pivô foi colocado
    return i + 1


# Função principal do QuickSelect que encontra o k-ésimo menor elemento
def quickselect(lista, esq, dir, k):
    if esq <= dir:      # verifica se ainda existe uma parte válida da lista para analisar

        # chama a função de partição para reorganizar a lista e obter a posição final do pivô
        pivo_index = particao(lista, esq, dir)

        # se a posição do pivô for exatamente k, encontramos o k-ésimo menor elemento
        if pivo_index == k:
            return lista[pivo_index]

        # se a posição do pivô for maior que k, então o elemento procurado está na parte esquerda da lista
        elif pivo_index > k:
            return quickselect(lista, esq, pivo_index - 1, k)
        else:                                                       # caso contrário, o elemento está na parte direita da lista
            return quickselect(lista, pivo_index + 1, dir, k)

    return None         # caso a busca não encontre o elemento (situação limite)
```
