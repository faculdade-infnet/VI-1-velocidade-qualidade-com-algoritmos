# 2 - Interseção de Arrays em Tempo Linear

## 2.1 - Class Hash Table

```python
class HashTable:
    # O(n) pois cria "n" listas vazias
    def __init__(self, tamanho):
        self.tamanho = tamanho
        self.tabela = [[] for _ in range(tamanho)]  # cria o numero(tamanho) de listas vazias solicitadas

    # transforma a chave em um índice válido do array (45 % 10 = 5 posição)
    def _funcao_hash(self, chave):
        return chave % self.tamanho

    # insere um novo valor caso ele não exista, se existir atualiza
    def inserir(self, chave, valor):
        # calcula o índice
        index= self._funcao_hash (chave)
        bucket = self.tabela [index]

        # verificar se a chave já existe e atualização do valor
        for i, (chave_existente, _) in enumerate (bucket):
            if chave_existente == chave:
                bucket[i] = (chave, valor)
                return
        # se a chave não existe, adicionar o novo par chave-valor no bucket
        bucket.append((chave, valor))

    # busca um valor dentro do bucket/tabela
    def buscar(self, chave):
        # calcula o índice
        index= self._funcao_hash (chave)
        bucket = self.tabela [index]
        # procurar pela chave e retornar o valor correspondente
        for chave_existente, valor in bucket:
            if chave_existente == chave:
                return valor

        return None # chave não encontrada!
```

```python
# Popular array na Hash Table
def populateHashTable(array):
    hash_table = HashTable(10)

    for item in array:
        if hash_table.buscar(item) is None:
            hash_table.inserir(item, 1)
        else:
            valor = hash_table.buscar(item)
            hash_table.inserir(item, valor + 1)

    return hash_table
```

## 2.2 - Function - Intersecção

```python
# Encontra os valores comuns entre os dosi arrays
def intersection(array1, array2):
    hash_table = populateHashTable(array1)    # primeiro array, pupular
    elementos_comuns = []                     # segundo array
    print(hash_table.tabela)

    for i in array2:
        valor = hash_table.buscar(i)
        if valor is not None and valor > 0:
            elementos_comuns.append(i)
            hash_table.inserir(i, 0)  # marca como já usado, evitando duplicata

    elementos_comuns.sort()
    return elementos_comuns
```

```python
import random
# Testando

# -------- GERAR LISTAS ALEATÓRIAS 5 A 20 NÚMEROS DO Nº 1 AO 100 --------
listas = []

for _ in range(6):
    tamanho = random.randint(50000, 100000)                   # define o número de itens na lista 5 a 20 itens
    lista = [random.randint(1, 1000) for _ in range(tamanho)]   # criar lista com números de 1 a 10 (podendo repetir)
    listas.append(lista)                                      # adiciona a lista a listas(com 5 listas)

# -------- PERCORRE LISTAS E IMPRIME O VALORES EM COMUM --------
for lista1, lista2 in zip(listas, listas[1:]):
    elementos_comuns = intersection(lista1, lista2)

    print(f'Lista 1: {lista1}')
    print(f'Lista 2: {lista2}')
    lista_planificada = ", ".join(str(x) for x in elementos_comuns)
    print(f'Elementos em comum: {lista_planificada}\n')
```
