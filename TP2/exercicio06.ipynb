# 1 - Selection Sort

```python
import time

def selection_sort(lista):
    n = len(lista)
    comparacoes = 0 # número de comparações executadas
    trocas = 0      # número de trocas executadas

    # percorre a lista para localizar o menor elemento
    for i in range(n):
        min_index = i   # assume que o elemento atual é o menor

        # encontramos o menor elemento na parte não ordenada da lista
        for j in range(i + 1, n):   # Start index, Stop index
            comparacoes += 1
            # Se estiver fora de ordem → troca.
            if lista[j] < lista[min_index]:
                min_index = j   # Atualiza índice do menor elemento

        # trocando o menor elemento com o primeiro elemento da parte não ordenada
        if min_index != i:
            trocas += 1
            lista[i], lista[min_index] = lista[min_index], lista[i]

    print(f'\tNúmero comparacoes: {comparacoes}')
    print(f'\tNúmero trocas: {trocas}')

# testando
listas = [
    ("lista aleatoria", [47, 12, 89, 3, 76, 54, 28, 91, 65, 10, 38, 72, 5, 99, 41, 23, 84, 16, 58, 67, 2, 95, 31, 44, 79, 18, 60, 7, 88, 52]),
    ("lista ordenada", list(range(1, 60, 2))),
    ("lista invertida", list(range(60, 0, -2))),
    ("lista pequena", [21, 28, 16, 24, 6, 3, 27, 22, 29, 1]),
    ("lista grande ordenada", list(range(60, 0, -2)))
]

for i, (nome, lista) in enumerate(listas, start=1):
    print(f"{i} - {nome}:")
    inicio_t = time.perf_counter()
    selection_sort(lista)
    total_linear = time.perf_counter() - inicio_t
    print(f"\tTempo: {total_linear:.5f} segundos = {lista}")
```
