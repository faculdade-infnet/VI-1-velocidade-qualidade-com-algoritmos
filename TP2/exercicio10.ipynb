# 10 - Fila com Ponteiros

```python
class Fila:
    # cria uma nova fila vazia
    def __init__(self, capacidade):
        self.capacidade = capacidade        # adicionar atributo ao objeto
        self.itens = [None] * capacidade
        self.inicio = 0
        self.fim = 0
        self.tamanho = 0

    # verifica se a fila está vazia
    def vazia(self):
        return self.tamanho == 0

    # adiciona um novo item no final da fila (push)
    def enfileirar (self, item):
        # Se fila = capacidade máxima de itens, informa que está cheia
        if self.tamanho == self.capacidade:
            print("Overflow! Fila cheia.")
            return
        else:
            self.itens[self.fim] = item
            self.fim = (self.fim + 1) % self.capacidade
            self.tamanho += 1

    # remove o último da fila (pop)
    def desenfileirar (self):
        # Se fila = vazia, informa que não pode remover itens
        if self.tamanho == 0:
            print("Underflow! Não é possível remover: fila vazia.")
            return None
        else:
            item = self.itens[self.inicio]
            self.itens[self.inicio] = None
            self.inicio = (self.inicio + 1) % self.capacidade
            self.tamanho -= 1
            return item

    # retorna o primeiro item na fila
    def primeiro(self): # top
        if self.tamanho == 0:
            return None
        else:
            return self.itens[self.inicio]

    # retorna o numero de elementos na fila
    def tamanho_fila(self):
        return self.tamanho

    # esvaziar fila, remove todos os dados
    def esvaziar(self):
        self.itens = [None] * self.capacidade
        self.inicio = 0
        self.fim = 0
        self.tamanho = 0

    # imprime todos os itens existentes na fila
    def imprimir(self):
        print(f"\t{self.itens} | inicio: {self.inicio} | fim: {self.fim} | tamanho: {self.tamanho}")
```

```python
# Testando a FILA
fila = Fila(capacidade=5)
fila.imprimir()
cont = 0

for i in range(4, 13, 2):
    cont += 1
    fila.enfileirar(i)
    print(f'{cont}-Enfileirar (adiciona o item {i})' )
    print(f"\tPrimeiro item:", fila.primeiro())
    fila.imprimir()

item_removido = fila.desenfileirar()
print(f"6-Desenfileirar (remove item {item_removido}):")
fila.imprimir()
print(f"7-Primeiro item:", fila.primeiro())
fila.imprimir()

item_removido = fila.desenfileirar()
print(f"8-Desenfileirar (remove item {item_removido}):")
fila.imprimir()
print(f"9-Primeiro item:", fila.primeiro())
fila.imprimir()

item = 4
fila.enfileirar(item)
print(f'10-Enfileirar (adiciona o item {item})')
fila.imprimir()
print(f"11-Primeiro item:", fila.primeiro())
fila.imprimir()
```
