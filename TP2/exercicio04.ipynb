# 4 - Letra ausente no alfabeto

```python
import string

def missing_letter(texto):
    # cria o alfabeto de "a" até "z", onde por default tudo é False
    alfabeto = [[chr(i), False] for i in range(ord('a'), ord('z') + 1)]

    # i = letra
    for i in texto:
        if 'a' <= i <= 'z':               # garante letras de 'a' até 'z'
            indice = ord(i) - ord('a')    # localiza o índice da letra no alfabeto
            # print(indice, "-", i)
            alfabeto[indice][1] = True    # Marca que localizou a letra no alfabeto

    letras_encontradas = []
    letras_nao_encontradas = []
    for i in alfabeto:
        # verifica se a segunda do sub array no alfabeto = False (letra não foi encontrada)
        if i[1] == False:
            # adiciona letra não econtrada na nova lista
            letras_nao_encontradas.append(i[0])
        else:
            letras_encontradas.append(i[0])

    return(letras_encontradas, letras_nao_encontradas)

# Testando
#texto = "hoje é um dia de sol que Parece não acabar mais"
texto = "acdefghijklmnopqrstuvyxZw"
#texto = "a D , . #"
(letras_encontradas, letras_nao_encontradas) = missing_letter(texto)
result_encontradas = ", ".join(letras_encontradas)
result_nao_encontradas = ", ".join(letras_nao_encontradas)
print(f"Letras faltando: ", result_nao_encontradas)
print(f"Letras encontradas: ", result_encontradas)
```
