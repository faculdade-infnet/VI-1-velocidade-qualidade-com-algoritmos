# 11 - Pilha reversão de String

```python
class ReverterString:
    # cria uma nova pilha vazia com capacidade máxima
    def __init__(self, capacidade):
        self.capacidade = capacidade        # adicionar atributo ao objeto
        self.itens = []

    # verifica se a pilha está vazia
    def vazia(self):
        return not self.itens

    # adiciona um novo item no topo da pilha (push)
    def empilhar (self, item):
        # Se pilha = capacidade máxima de itens, informa que está cheia
        if len(self.itens) == self.capacidade:
            print("Overflow! Pilha cheia.")
            return
        else:
            self.itens.append(item)

    # remove o item do top e retorna o item removido (pot)
    def desempilhar (self):
        # Se pilha = vazia, informa que não pode remover itens
        if self.vazia():
            print("Underflow! Não é possível remover: pilha vazia.")
            return None
        else:
            return self.itens.pop()

    # imprime todos os itens existentes na pilha
    def imprimir_apos_empilhar (self):
        resultado = "".join(self.itens)
        print(f"String após empilhar: {resultado}")
```

```python
# Testando a PILHA - com diferentes strings
listas = [
    ["ABCDEF"],
    ["Pythonic"],
    ["StackClass"],
    ["DataStruct12"],
    ["ReverseStackABC"]
]

for item in listas:
    item = item[0]  #trasforma em string
    pilha = ReverterString(capacidade=len(item))

    print(f"\nOriginal: {pilha.itens}")
    for i, letra in enumerate(item):
        pilha.empilhar(letra)

    pilha.imprimir_apos_empilhar()

    string_invertida = ""
    while not pilha.vazia():
        letra_removida = pilha.desempilhar()
        string_invertida += letra_removida

    print(f"String invertida: {string_invertida}")
```
