# 1 - Otimização de Código Quadrático com Hash Table

## 1.1- Function - O(n²) Maior número com dois laços aninhados

```python
# Encontrar o maior número único
def greatestNumber(array):
    comparacoes = 0
    maiorValorUnico = None

    for i in array:
        repeticoes = 0
        # Encontra dos valore duplicados comparando cada i com todos j
        for j in array:
            # Se encontrarmos outro valor maior que i, então, i não é o maior:
            comparacoes += 1
            if j == i:
               repeticoes += 1

        # Se aparece apenas uma vez
        if repeticoes == 1:
            if maiorValorUnico is None or i > maiorValorUnico:
                maiorValorUnico = i

    return maiorValorUnico, comparacoes
```

## 1.2 - Class + Function - O(n) Maior número com Hash Table

```python
class HashTable:
    # O(n) pois cria "n" listas vazias
    def __init__(self, tamanho):
        self.tamanho = tamanho
        self.tabela = [[] for _ in range(tamanho)]  # cria o numero(tamanho) de listas vazias solicitadas

    # transforma a chave em um índice válido do array (45 % 10 = 5 posição)
    def _funcao_hash(self, chave):
        return chave % self.tamanho

    # insere um novo valor caso ele não exista, se existir atualiza
    def inserir(self, chave, valor):
        # calcula o índice
        index= self._funcao_hash (chave)
        bucket = self.tabela [index]

        # verificar se a chave já existe e atualização do valor
        for i, (chave_existente, _) in enumerate (bucket):
            if chave_existente == chave:
                bucket[i] = (chave, valor)
                return
        # se a chave não existe, adicionar o novo par chave-valor no bucket
        bucket.append((chave, valor))

    # busca um valor dentro do bucket/tabela
    def buscar(self, chave):
        # calcula o índice
        index= self._funcao_hash (chave)
        bucket = self.tabela [index]
        # procurar pela chave e retornar o valor correspondente
        for chave_existente, valor in bucket:
            if chave_existente == chave:
                return valor

        return None # chave não encontrada!
```

```python
# Encontrar o maior número único com Hash Table --------
def greatestNumberHashTable(array):
    hash_table = HashTable(len(array))
    maiorValorUnico = None
    acessos = 0

    # Inserir todos itens, contanto frequencia (i= numero)
    for i in array:
        acessos += 1
        numero_encontrado = hash_table.buscar(i)            # busca o numero na hash table
        if numero_encontrado is None:
            acessos += 1
            hash_table.inserir(i, 1)                        # insere o numero se ele não foi encontrado
        else:
            acessos += 1
            hash_table.inserir(i, numero_encontrado + 1)    # atualiza o contador de ocorrencias do número

    print(hash_table.tabela)

    # Percorrer todos os buckets da hash_table
    for bucket in hash_table.tabela:
        acessos += 1
        for chave, valor in bucket:
            acessos += 1
            # se a contagem for =1, número único
            if valor == 1:
                if maiorValorUnico is None or chave > maiorValorUnico:
                    maiorValorUnico = chave

    return maiorValorUnico, acessos
```

```python
import random
# Testando

# -------- GERAR LISTAS ALEATÓRIAS 5 A 20 NÚMEROS DO Nº 1 AO 100 --------
listas = []

for _ in range(5):
    tamanho = random.randint(5, 20)                           # define o número de itens na lista 5 a 20 itens
    lista = [random.randint(1, 10) for _ in range(tamanho)]   # criar lista com números de 1 a 10 (podendo repetir)
    listas.append(lista)                                      # adiciona a lista a listas(com 5 listas)

# -------- PERCORRE LISTAS E IMPRIME O MAIOR NÚMERO --------
for lista in listas:
    maiorValorUnico, comparacoes = greatestNumber(lista)
    print(f"### Tamanho: {len(lista)} ###")
    print("----- LAÇOS ANINHADOS -----")
    print(f'Número comparacoes: {comparacoes}')
    print(f'O maior valor é {maiorValorUnico} na lista: {lista}')

    print("----- HASH TABLE -----")
    maiorValorUnico, acessos = greatestNumberHashTable(lista)
    print(f'Número de acessos: {acessos}')
    print(f'O maior valor é {maiorValorUnico} na lista: {lista}\n')
```
