# 9 - Pilha com Controle de Erros

```python
class Pilha:
    # cria uma nova pilha vazia com capacidade máxima
    def __init__(self, capacidade):
        self.capacidade = capacidade        # adicionar atributo ao objeto
        self.itens = []

    # verifica se a pilha está vazia
    def vazia(self):
        return not self.itens

    # adiciona um novo item no topo da pilha (push)
    def empilhar (self, item):
        # Se pilha = capacidade máxima de itens, informa que está cheia
        if len(self.itens) == self.capacidade:
            print("Overflow! Pilha cheia.")
            return
        else:
            self.itens.append(item)

    # remove o item do top e retorna o item removido (pot)
    def desempilhar (self):
        # Se pilha = vazia, informa que não pode remover itens
        if self.vazia():
            print("Underflow! Não é possível remover: pilha vazia.")
            return None
        else:
            return self.itens.pop()


    # retorna o item do topo da pilha
    def topo(self): # top
        if not self.vazia():
            return self.itens[-1]

    # retorna o numero de elementos da pilha
    def tamanho(self):
        return len(self.itens)

    # esvaziar pilha, remove todos os dados
    def esvaziar(self):
        self.itens.clear()

    # imprime todos os itens existentes na pilha
    def imprimir (self):
        print(f"\t",self.itens)
```

```python
# Testando a PILHA
entrada = [2, 4, 6, 8, 10]
pilha = Pilha(capacidade=5)
for i in entrada:
    pilha.empilhar(i)
    pilha.imprimir()


print("1-Item no Topo da pilha:", pilha.topo())
pilha.imprimir()
print("2-Tamanho da pilha:", pilha.tamanho())
pilha.imprimir()
print("3-Desempilhar (remove item do topo):", pilha.desempilhar())
pilha.imprimir()

item = 12
pilha.empilhar(item)
print(f'4-Empilhar (adiciona o item {item} no topo da pilha)')
pilha.imprimir()
print("5-Item no Topo da pilha:", pilha.topo())
pilha.imprimir()
print("6-Tamanho da pilha:", pilha.tamanho())
pilha.imprimir()

item = 14
pilha.empilhar(item)
print(f'7-Empilhar (adiciona o item {item} no topo da pilha)')
pilha.imprimir()
print("8-Item no Topo da pilha:", pilha.topo())
pilha.imprimir()
print("9-Tamanho da pilha:", pilha.tamanho())

pilha.esvaziar()
pilha.imprimir()
print("10-Tamanho da pilha:", pilha.tamanho())
pilha.imprimir()
pilha.desempilhar()
pilha.imprimir()
```
