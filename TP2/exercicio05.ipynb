# 5 - Primeiro caractere não repetido

```python
# contar frequencia em array com sub array = [[a, 1,], [b, 2], [c, ] ]
# para encontrar o primei caractere não repetido, após o for acim, percorrer a lista e enocntrar onde a segunda posição[1] = [caractere, 1]
import string

def first_letter(texto):
    # conta as ocorrências das letras encontradas {caractetere, contador}
    caracteres_contagem = {}

    # i = letra
    for i in texto:
        if i in caracteres_contagem:
            caracteres_contagem[i] += 1
        else:
            caracteres_contagem[i] = 1

    # armazena o primeiro caractere não repetido
    primeiro_caractere = "";
    for letra, i in caracteres_contagem.items():
        # verifica o primeiro valor = 1 pra o sub array no alfabeto (primeira letra encontrada)
        if i == 1:
            # atribui letra econtrada
            primeiro_caractere = letra
            break

    return(primeiro_caractere, caracteres_contagem)


texto1 = "aB3! xYz9@#LmN0$pQ7kL& tR2*e"
texto2 = "W8+vZ1^mN4% pQ6!sT9@uV2#Xy1$"
texto3 = "zA3&bC5*dE7!9fG@ hJ2#kL4%mN"
texto4 = "6&Op8( qR0)St2=Uv4?Wx6/Yz8"
texto5 = "Ab1!Cd3@Ef5#Gh7$Ij9%Kl0^Mn"
lista = [texto1, texto2, texto3, texto4, texto5]

for i, texto in enumerate(lista):
    i += 1
    (primeiro_caractere, caracteres_contagem) = first_letter(texto)
    print(f"Primeiro caractere não duplicado L{i}:", primeiro_caractere)
    print(f"Contagem de caracteres L{i}:\n", caracteres_contagem)
    print()
```
