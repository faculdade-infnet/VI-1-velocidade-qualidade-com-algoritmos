## Busca Sequencial

```python
# Busca Sequencial (ou Busca Linear):
lista = ["Pedro", "Mateus", "João", "André", "Felipe", "Tadeu"]
b = False
nome = input("Nome a ser localizado na lista: ")

for i in range(len(lista)):
  if lista[i] == nome:
    b = True
    print (nome, "é o " + str(i + 1) + " elemento da lista")
    break
if b == False:
  print(nome, "não consta na lista")
```

## Busca Binária

```python
# Busca Binaria:
def busca_binaria(lista, nome):
  inicio = 0
  fim = len(lista) - 1

  while inicio <= fim:
    meio = (inicio + fim)//2   # divisão inteira
    if lista[meio] == nome:
      return meio
    if lista[meio] > nome:
      fim = meio - 1
    else:
      inicio = meio + 1
  return -1


# exemplo:
lista = ["Pedro", "Mateus", "João", "André", "Felipe", "Tadeu"]
lista.sort()
print("Lista ordenada: ", lista)
nome = input("Nome a ser localizado na lista: ")
pos = busca_binaria(lista, nome)

if pos == -1:
  print(nome, "não consta na lista")
else:
  print (nome, "é o " + str(pos + 1) + "°- elemento da lista")
```

## Comparação Busca Sequencial x Busca Binária

```python
# Comparação Busca Sequencial x Busca Binária
import time   # relógio de alta precisão, ideal para benchmarks
import random # gerar números aleatórios

# função para a busca linear: # procura item percorrendo a lista inteira
def busca_linear(lista, item):
  for i in range(len(lista)):
    # se item atual da lista == item a ser buscado, então retorna o indice do item
    if lista[i] == item:
      return i
  return -1


# função para a busca binária:
def busca_binaria (lista, item):
  # len(lista) - 1, pois a lisa inicia em 0
  inicio, fim = 0, len(lista) - 1

  # divide a lista enquanto ainda existir um intervalo válido
  while inicio <= fim:
    # econtra o indice central
    meio = (inicio + fim)//2

    # se for igual retorna o itemm, senão continua a divisão
    if lista[meio] == item:
      return meio
    if lista[meio] > item:
      fim =  meio -1
    else:
      inicio = meio +1
  return -1


# configurações de teste:
num_testes = 10000    # quantas vezes o teste será repetido
tamanho_lista = 1000  # tamanho da lista em cada teste
total_linear = 0      # soma do tempo da busca linear
total_binaria = 0     # soma do tempo da busca binária

# Execução dos testes:
# _ = sem valor no contador, só repete o teste
for _ in range(num_testes):
  # criar listas aleatórias ordenadas
  lista = sorted([random.randint(0, 10000) for _ in range(tamanho_lista)])
  # numero a ser procurado
  alvo = random.randint(0, 10000)

  # medição com busca linear:
  # Executa a busca, Calcula o tempo gasto, Soma ao total
  inicio_t = time.perf_counter()                 # marca o tempo inicial
  busca_linear (lista, alvo)                     # executa a busca
  total_linear += time.perf_counter() - inicio_t # calcula o tempo gasto, Soma ao total

  # medição com busca binária, logica igual ao anterior
  inicio_t = time.perf_counter()
  busca_binaria(lista, alvo)
  total_binaria += time.perf_counter() - inicio_t

# verifica qula busca foi mais rapida no total
if total_linear > total_binaria:
  busca = "Busca Binaria"
  mult = total_linear/total_binaria   # quantas vezes foi mais rápida [%]
else:
  busca = "Busca Linear"
  mult = total_binaria/total_linear   # quantas vezes foi mais rápida [%]

# Resultados
print (f"Resultados para {num_testes} testes: ")
print (f"Tempo total da busca linear: {total_linear:.5f} segundos")
print (f"Tempo total da busca binária: {total_binaria:.5f} segundos")
print (f"A busca {busca} foi {mult:.2f} vezes mais rápida")
```
