# Recursividade



### 1 - Fatorial

```python
# "n" representa o número que queremos calcular o fatorial
def fatorial(n):
    # CASOS BASE, faz a função parar de chamar ela mesma, matemáticamente: (0! = 1) e (1! = 1)
    if n == 0 or n ==1:
        return 1            # retorna 1, porque fatorial de (0 e 1) = 1
    else:
        # aqui acontece a recursividade: a função chama ela mesma, quebrando o problema em um menor
        # multiplicamos n pelo fatorial de (n-1)
        return n * fatorial(n-1)

# TESTANDO
# laço vai de 0 a 19 = 20 vezes
for i in range(20):
    # para cada número de 0 até 19 ele calcula o fatorial
    # imprime o número(i) e o resultado fatorial de "i"
    print(i,  fatorial(i))
```

### 2 - Fibonacci

```python
# Recusividade Direta - "n" representa a posição na sequência de Fibonacci que queremos calcular
def fibonacci(n):
    # CASOS BASE da recursividade, na sequência de Fibonacci: F(1) e F(2) = 1
    if n == 1 or n == 2:
        return 1             # retorna 1 porque os dois primeiros valores da sequência são 1
    else:
        # valor atual é a soma dos dois anteriores
        # função chama ela mesma duas vezes, p/ calcular F(n-1) e p/ calcular F(n-2)
        return fibonacci(n-1) + fibonacci(n-2)

# TESTANDO
# laço vai de 1 a 10 = 10 vezes
for i in range(1, 11):
    # imprime o número(i) e o valor de Fibonacci em "i"
    print(i, fibonacci(i))
```

### 3 - Fibonacci Interativo, com variáveis

```python
# COM VARIÁVEIS, pois o cálculo é feito guardando os valores em variáveis (f1 = f2 = 1)
# "n" representa a posição na sequência de Fibonacci que queremos calcular
def fibonacci2(n):
    # CASOS BASE da recursividade, na sequência de Fibonacci: F(1) e F(2) = 1
    if n == 1 or n == 2:
        return 1             # retorna 1 porque os dois primeiros valores da sequência são 1
    else:
        # criamos duas variáveis para guardar os dois últimos valores da sequência
        # f1 representa F(n-2) --> F(1) = 1
        # f2 representa F(n-1) --> F(2) = 1
        f1 = f2 = 1
        for i in range(3, n + 1):     # calcular os próximos valores da sequência, inicio em 3 porque os dois primeiros já sabemos
            # Atualiza os dois últimos valores da sequência
            # f1 recebe o valor antigo de f2
            # f2 recebe a soma dos dois anteriores (próximo Fibonacci)
            f1, f2 = f2, f1 + f2
        return f2  # f2 contém o Fibonacci de n

# TESTANDO
# laço vai de 1 a 10 = 10 vezes
for i in range(1, 11):
    # imprime o número(i) e o valor de Fibonacci em "i"
    print(i, fibonacci2(i))
```

### 4 - Fibonacci Interativo, com lista

```python
# COM LISTA, pois usa ela para guardar todos os valores da sequência
# "n" representa a posição na sequência de Fibonacci que queremos calcular
# O cálculo é feito aumntando "i" até atingir "n"
def fibonacci3(n):
    # Cria uma lista com os dois primeiros valores da sequência de Fibonacci
    lista = [0, 1]

    # inica em 2, pois 0 e 1 já existem na lista
    for i in range (2, n+1):
        # Calcula o próximo valor da sequência, pegando os dois valores anteriores da lista
        # lista[i-1] → Fibonacci penultimo
        # lista[i-2] → Fibonacci antepenúltimo
        # soma os dois valores e adiciona no final da lista
        lista.append(lista[i-1] + lista[i-2])

    return lista[n], lista  # quando chegar em "n", retorna o ultimo valor da lista

# TESTANDO
# obtem o Fibonacci da posição 10
item, lista = fibonacci3(20)
print(f'{item}\nLista gerada {lista}')
```

# Torre de Hanoi

```python
# n = número de discos
# origem = pino onde os discos começam
# destino = pino para onde queremos mover os discos
# auxiliar = pino usado temporariamente para ajudar na movimentação
def torre_hanoi(n, origem, destino, auxiliar):

    # CASO BASE da recursão - se existir apenas 1 disco, basta mover ele direto
    if n == 1:
        print(f"Mover disco {n} da torre {origem} para torre {destino}")
        return

    # PASSO 1
    # mover os n-1 discos de cima da torre
    # da origem para o pino auxiliar
    # usando o destino como apoio
    torre_hanoi(n-1, origem, auxiliar, destino)

    # PASSO 2
    # mover o maior disco (o disco n)
    # da origem para o destino
    print(f"Mover disco {n} da torre {origem} para torre {destino}")

    # PASSO 3
    # mover os n-1 discos que estavam no auxiliar
    # para o destino
    # usando a origem como apoio
    torre_hanoi(n-1, auxiliar, destino, origem)


# TESTE
# resolver a torre com 3 discos
# 3 discos A=origem | C=destino |B=auxiliar
torre_hanoi(3, "A", "C", "B")

# Inicio
#          |  1  |     |     |        |     |     |     |        |     |     |     |
#          |  2  |     |     |  --->  |  2  |     |     |  --->  |     |     |     |
#          |  3  |     |     |        |  3  |     |  1  |        |  3  |  2  |  1  |
# torres = |  A  |  B  |  C  |        |  A  |  B  |  C  |        |  A  |  B  |  C  |

# Objetivo
#          |     |     |  1  |
#          |     |     |  2  |
#          |     |     |  3  |
# torres = |  A  |  B  |  C  |
```

# Problema de Josephus - Recursivo

```python
# n = número de pessoas
# k = passo de eliminação
# A função retorna a posição do sobrevivente (índice começando em 0)
def josephus_recursivo(n, k):

    # CASO BASE
    # Se existir apenas 1 pessoa, ela é o sobrevivente
    if n == 1:
        return 1

    # PASSO RECURSIVO
    # Resolve primeiro o problema menor (n - 1, k)
    # Ajusta a posição porque a contagem anda k passos (+ k - 1 ) pois iniciamos em 1
    # (% n) faz a contagem "voltar para o início" quando passa do final da lista.
    # (+ 1) converte o resultado para posições humanas (1 até n)
    # Isso simula o CÍRCULO de pessoas.
    return (josephus_recursivo(n - 1, k) + k -1) % n + 1


# TESTANDO
n = 7 # pessoas
k = 3 # (a cada 3 pessoas uma é eliminada)

sobrevivente = josephus_recursivo(n, k)
print(f"\nNúmero de pessoas {n}, salto a cada {k} pessoas")
print("Sobrevivente:", sobrevivente)

# Sequência
# [1, 2, 3, 4, 5, 6, 7] = Contagem: 1 → 2 → 3 = 3 é eliminado ❌
# [1, 2, 4, 5, 6, 7]    = Contagem: 4 → 5 → 6 = 6 é eliminado ❌
# [1, 2, 4, 5, 7]       = Contagem: 7 → 1 → 2 = 2 é eliminado ❌
# [1, 4, 5, 7]          = Contagem: 4 → 5 → 7 = 7 é eliminado ❌
# [1, 4, 5]             = Contagem: 1 → 4 → 5 = 5 é eliminado ❌
# [1, 4]                = Contagem: 1 → 4 → 1 = 1 é eliminado ❌
# [4]✅ SOBREVIVENTE FINAL
```

# Problema de Josephus - Iterativo

```python
# n = número de pessoas
# k = passo de eliminação
# A função retorna a posição do sobrevivente (índice começando em 0)
def josephus_iterativo(n, k):
    pessoas = list(range(1, n + 1))   # lista de pessoas de 1 até n
    indice = 0                        # posição inicial, indica de onde começa a contagem no círculo

    # Enquanto existir mais de uma pessoa no círculo
    while len(pessoas) > 1:
        # Calcula quem será eliminado
        # indice + k - 1 → anda k posições a partir da posição atual
        # % len(pessoas) → faz a contagem voltar para o início da lista
        # simulando o círculo
        indice = (indice + k - 1) % len(pessoas)
        pessoas.pop(indice)                      # elinmina uma pessoa

    # Quando o loop termina sobra apenas uma pessoa, entãao retorna ela
    return pessoas[0]


# TESTANDO
n = 7 # pessoas
k = 3 # (a cada 3 pessoas uma é eliminada)

sobrevivente = josephus_iterativo(n, k)
print(f"\nNúmero de pessoas {n}, salto a cada {k} pessoas")
print("Sobrevivente:", sobrevivente)

# Sequência
# [1, 2, 3, 4, 5, 6, 7] = Contagem: 1 → 2 → 3 = 3 é eliminado ❌
# [1, 2, 4, 5, 6, 7]    = Contagem: 4 → 5 → 6 = 6 é eliminado ❌
# [1, 2, 4, 5, 7]       = Contagem: 7 → 1 → 2 = 2 é eliminado ❌
# [1, 4, 5, 7]          = Contagem: 4 → 5 → 7 = 7 é eliminado ❌
# [1, 4, 5]             = Contagem: 1 → 4 → 5 = 5 é eliminado ❌
# [1, 4]                = Contagem: 1 → 4 → 1 = 1 é eliminado ❌
# [4]✅ SOBREVIVENTE FINAL
```
