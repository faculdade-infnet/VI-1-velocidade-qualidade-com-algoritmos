# Pilhas

```python
class Pilha:
    # cria uma nova pilha vazia
    def __init__(self):
        self.itens = []

    # verifica se a pilha está vazia
    def vazia(self):
        return not self.itens

    # adiciona um novo item no topo da pilha na veritcal
    def empilhar (self, item): # push
        self.itens.append(item)

    # remove o último item inserido do top e retorna o item removido da pilha na veritcal
    def desempilhar (self): # pop
        if not self.vazia():
            return self.itens.pop()
        else:
            return None

    # retrona o item do topo na vertical
    def topo(self): # top
        if not self.vazia():
            return self.itens[-1]

    # retorna o numero de elementos da lista na vertical
    def tamanho(self):
        return len(self.itens)

    # imprime todos os itens existentes na lista atual
    def imprimir (self):
        print(self.itens)
```

```python
# Testando a PILHA
entrada = [2, 4, 6, 8, 10]
pilha = Pilha()
for i in entrada:
    pilha.empilhar(i)


pilha.imprimir()
print("Item no Topo da pilha:", pilha.topo())
print("Tamanho da pilha:", pilha.tamanho())
print("Desempilhar (remove item do topo):", pilha.desempilhar())
pilha.imprimir()
print("Item no Topo da pilha:", pilha.topo())
print("Tamanho da pilha:", pilha.tamanho())

item = 12
pilha.empilhar(item)
print(f'Empilhar (adiciona o item {item} no topo da pilha)')
pilha.imprimir()
print("Item no Topo da pilha:", pilha.topo())
print("Tamanho da pilha:", pilha.tamanho())
```

# Filas

```python
class Fila:
    # cria uma nova fila vazia
    def __init__(self):
        self.itens = []

    # verifica se a fila está vazia
    def vazia(self):
        return self.itens == []

    # adiciona um novo item na fila na horizontal
    def enfileirar (self, item): # push
        self.itens.append(item)

    # remove o primeiro item inserido na fila na horizontal
    def desenfileirar (self): # pop
        if not self.vazia():
            return self.itens.pop(0)
        else:
            return None

    # retorna o primeiro item na horizontal
    def primeiro(self): # top
        if not self.vazia():
            return self.itens[0]
        else:
            return None

    # retorna o numero de elementos da lista na horizontal
    def tamanho(self):
        return len(self.itens)

    # imprime todos os itens existentes na lista atual
    def imprimir (self):
        print(self.itens)
```

```python
# Testando a FILA
entrada = [4, 6, 8, 10, 12]
fila = Fila()
for i in entrada:
    fila.enfileirar(i)


fila.imprimir()
print("Primeiro item:", fila.primeiro())
print("Tamanho da fila:", fila.tamanho())
print("Desenfileirar (remove o primeiro item inserido):", fila.desenfileirar())
fila.imprimir()
print("Primeiro item:", fila.primeiro())
print("Tamanho da fila:", fila.tamanho())

item = 14
fila.enfileirar(item)
print(f'Enfileirar (adiciona o item {item} após o último item inserido)' )
fila.imprimir()
print("Primeiro item:", fila.primeiro())
print("Tamanho da fila:", fila.tamanho())
```
