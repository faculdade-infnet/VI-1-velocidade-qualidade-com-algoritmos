# 1. Exibe um quadrado visual de 10 x 10

```python
# Exibe um quadrado visual de 10 x 10
n = 101
crivo = [1] * n

for i in range(n):
  if i % 10 != 0:
    print(crivo[i], " ", end=" ")
  else:
    print(crivo[i])
```

# 2. Encontrar posição dos números primos [1]

```python
import math
# Na lista impressa
# 1 - significa "ainda pode ser primo"
# 0 - significa "não é primo"
n = 101
crivo = [1] * n
crivo[0] = crivo[1] = 0

# Número a ser testado
i = 2
# Eliminar múltiplos apenas dos números até √n, porque qualquer número composto maior já terá sido eliminado por um divisor menor.
while (i < math.sqrt(n)):
  # Se "i" já foi eliminado como múltiplo de outro número menor, ignora
  if crivo[i] == 1:
    # Se "i" não é primo seus múltiplos também não são
      for j in range (2 * i, n,  i):
        crivo[j] = 0
  i += 1

# Exibe o quadrante com 0=não_primos e 1=primos
for i in range(n):
  if i % 10 != 0:
    print(crivo[i], " ", end=" ")
  else:
    print(crivo[i])
```

# 3. Exibir números primos [1]

```python
import math
# Na lista impressa
# 1 - significa "ainda pode ser primo"
# 0 - significa "não é primo"
n = 101
crivo = [1] * n
crivo[0] = crivo[1] = 0

# Número a ser testado
i = 2
# Eliminar múltiplos apenas dos números até √n, porque qualquer número composto maior já terá sido eliminado por um divisor menor.
while (i < math.sqrt(n)):
  # Se "i" já foi eliminado como múltiplo de outro número menor, ignora
  if crivo[i] == 1:
    # Se "i" não é primo seus múltiplos também não são
      for j in range (2 * i, n,  i):
        crivo[j] = 0
  i += 1

# Exibe o quadrante com 0=não_primos e 1=primos
for i in range(n):
  if i % 10 != 0:
    print(crivo[i], " ", end=" ")
  else:
    print(crivo[i])

# Exibe os número primos
for i in range(n):
  if crivo[i] == 1:
    print(i, end=" ")
```
