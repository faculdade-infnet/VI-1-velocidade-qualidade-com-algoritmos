# 1 - Bubble Sort - 9 passos

### 1.1. Simples - 90 passos

```python
# Buble Sort - Implementação IngÊnua.
# Ao contar todos os passos dentor do J, estou contando todos os passos
def bubble_sort_1(lista):
  n = len(lista)
  cont =  0 # conta número de passos executados

  # Loop exteno, controla o número de passagend do algoritmo.
  for i in range(n):

    # Loop interno para controlar as comparaçoes entre elementos.
    for j in range(n - 1):    # Stop index
      # Se estiver fora de ordem → troca.
      if lista[j] > lista[j + 1]:
        # Troca os elementos
        lista[j], lista[j + 1] = lista[j + 1], lista[j]
      cont +=1

  print(f'Número de passos: {cont}')

# testando
lista = [64, 34, 25, 12, 22, 11, 90, 18, 8, 9]
bubble_sort_1(lista)
print(lista)
```

### 1.2. Simples [index final diminui a cada passagem] - 45 passos

```python
# Buble Sort - Implementação.
# Ao contar todos os passos dentor do J, estou contando todos os passos
def bubble_sort_2(lista):
  n = len(lista)
  cont =  0 # conta número de passos executados

  # Loop exteno, controla o número de passagend do algoritmo.
  for i in range(n):
    # Loop interno para controlar as comparaçoes entre elementos.
    for j in range(0, n -i -1):   # Start index, Stop index
      # Se estiver fora de ordem → troca.
      if lista[j] > lista[j + 1]:
        # Troca os elementos
        lista[j], lista[j + 1] = lista[j + 1], lista[j]
      cont +=1

  print(f'Número de passos: {cont}')

# testando
lista = [64, 34, 25, 12, 22, 11, 90, 18, 8, 9]
bubble_sort_2(lista)
print(lista)
```

### 1.3. Otimizada [lista ordenada] - 9 passos

```python
# Buble Sort - Implementação.
# Ao contar todos os passos dentor do J, estou contando todos os passos
def bubble_sort_otimizado(lista):
  n = len(lista)
  cont =  0 # conta número de passos executados

  # Loop exteno, controla o número de passagens do algoritmo.
  for i in range(n):
    trocou = False
    # Loop interno para controlar as comparações entre elementos.
    # Último índice válido para comparação
    # [n(numero de elementos na lista) - i(contador atual) - 1(lista inicia em 0)]
    for j in range(0, n -i -1):   # Start index, Stop index
      cont +=1
      # Se estiver fora de ordem → troca.
      if lista[j] > lista[j + 1]:
        # Troca os elementos
        lista[j], lista[j + 1] = lista[j + 1], lista[j]
        trocou = True
    if not trocou:
      break

  print(f'Número de passos: {cont}')

# testando
lista = [64, 34, 25, 12, 22, 11, 90, 18, 8, 9]
bubble_sort_otimizado(lista)
print(lista)

lista_2 = [8, 9, 11, 12, 18, 22, 25, 34, 64, 90]
bubble_sort_otimizado(lista_2)
print(lista_2)
```

# 2 - Selection Sort - 45 passos

```python
# Selection Sort - Implementação em uma única lista.
# Ao contar todos os passos dentro do J, estou contando todos os passos
def selection_sort(lista):
  n = len(lista)
  cont =  0 # conta número de passos executados

  # percorre a lista para localizar o menor elemento
  for i in range(n):
    # assume que o elemento atual é o menor
    min_index = i

    # encontramos o menor elemento na parte não ordenada da lista
    for j in range(i + 1, n):   # Start index, Stop index
      # Se estiver fora de ordem → troca
      if lista[j] < lista[min_index]:
        # Troca os elementos
        min_index = j
      cont +=1
    # trocando o menor elemento com o primeiro elemento da parte não ordenada
    if min_index != i:
      lista[i], lista[min_index] = lista[min_index], lista[i]

  print(f'Número de passos: {cont}')

# testando
lista = [64, 34, 25, 12, 22, 11, 90, 18, 8, 9]
selection_sort(lista)
print(lista)

lista_2 = [8, 9, 11, 12, 18, 22, 25, 34, 64, 90]
selection_sort(lista_2)
print(lista_2)
```

# 3 - Insertion Sort

```python
# Insertion Sort - Implementação em uma única lista.
def insertion_sort(lista):
  n = len(lista)
  cont =  0 # conta número de passos executados

  # percorre a lista a partir do segundo elemento
  # Start index, Stop index
  for i in range(1, n):
    # o elemento atual é o escolhido para inserção
    chave = lista[i]
    # inicia a comparação a partir do elemento anterior da lista
    j = i - 1

    # mover para direita os elementos da sub-lista ordenada que são maiores do que "chave"
    while (j >= 0) and (lista[j] > chave):
      cont +=1
      lista[j + 1] = lista[j]
      j -=1
    # insere chave na posição correta
    lista[j + 1] = chave
  print(f'Número de passos: {cont}')

# testando
lista = [64, 34, 25, 12, 22, 11, 90, 18, 8, 9]
insertion_sort(lista)
print(lista)

lista_2 = [8, 9, 11, 12, 18, 22, 25, 34, 64, 90]
insertion_sort(lista_2)
print(lista_2)
```
