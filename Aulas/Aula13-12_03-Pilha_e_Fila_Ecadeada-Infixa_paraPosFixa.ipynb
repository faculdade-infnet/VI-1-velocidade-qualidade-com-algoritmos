# 13.1. Pilha por Encadedamento

```python
# Classe que representa um nó da pilha
class No:
    def __init__(self, valor):
        self.valor = valor                     # valor dentro do nó
        self.prox = None                       # ponteiro para o próximo nó (inicia vazio)

# Classe que representa a pilha
class Pilha:
    def __init__(self):
        self.topo = None                       # topo da pilha
        self._tamanho = 0                      # tamanho da pilha

    # Verifica se pilha = vazia, retorna True
    def esta_vazia(self):
        return self.topo is None

    # Retorna o tamanho da pilha
    def __len__(self):
        return self._tamanho

    # Adiciona um novo nó no topo da pilha e
    # aponta o novo nó para o topo antigo
    def empilhar(self, valor):
        novo_no = No(valor)                    # Cria um novo nó com o valor informado
        novo_no.prox = self.topo               # O novo nó aponta como próximo nó sendo o topo
        self.topo = novo_no                    # O topo agora passa a ser o novo nó
        self._tamanho += 1                     # Aumenta o tamanho da pilha

    # Remove o nó do topo da pilha e
    # atualiza o valor do topo
    def desempilhar(self):
        if self.esta_vazia():                  # Verifica se a pilha está vazia
            return "Pilha Vazia"

        valor = self.topo.valor                # Guarda o valor do topo
        self.topo = self.topo.prox             # O topo passa a ser o próximo nó
        self._tamanho -= 1                     # Diminui o tamanho da pilha
        return valor                           # Retorna o valor removido

    # Retorna o valor do topo se existir
    def ver_topo(self):
        if self.topo is None:
            return "Pilha Vazia"
        else:
            return self.topo.valor

    # Remove todos os nós
    def destruir(self):
        while self.topo is not None:
            temp = self.topo            # Guarda o nó atual
            self.topo = self.topo.prox  # Move o topo para o próximo nó
            temp.prox = None            # Apaga a referência do nó

        self._tamanho = 0               # Zera o tamanho
```

```python
# TESTANDO ----------------------------------------------
minha_pilha = Pilha()
minha_pilha.empilhar("Livro A")
minha_pilha.empilhar("Livro B")
minha_pilha.empilhar("Livro C")
minha_pilha.empilhar("Livro D")

print(f"Tamanho da atual da pilha: {minha_pilha.__len__()}")
print(f"Topo da pilha: {minha_pilha.ver_topo()}")
print(f"Removido: {minha_pilha.desempilhar()}")
print(f"Tamanho da atualizado da pilha: {minha_pilha.__len__()}")
print(f"Novo topo da pilha: {minha_pilha.ver_topo()}")
print(f"Removido: {minha_pilha.desempilhar()}")
print(f"Removido: {minha_pilha.desempilhar()}")
print(f"Removido: {minha_pilha.desempilhar()}")
print(f"Removido: {minha_pilha.desempilhar()}")
```

# 13.2. Fila por Encadedamento

```python
# Classe que representa um nó da fila -------------------------------------
class No:
    def __init__(self, valor):
        self.valor = valor                     # Guarda o valor dentro do nó
        self.prox = None                       # Ponteiro para o próximo nó (inicia vazio)

# Classe que representa a fila -------------------------------------
class Fila:
    def __init__(self):
        self.inicio = None                     # Aponta para o primeiro elemento da fila
        self.fim = None                        # Aponta para o último elemento da fila
        self._tamanho = 0                      # Guarda o tamanho da fila

    # Verifica se filha = vazia, retorna True
    def esta_vazia(self):
        return self.inicio is None

    # Retorna o tamanho da fila
    def __len__(self):
        return self._tamanho

    # Método para inserir elemento na fila
    def enfileirar(self, valor):
        novo_no = No(valor)                    # Cria um novo nó com o valor
        if self.esta_vazia():                  # Verifica se a fila está vazia
            self.inicio = self.fim = novo_no   # Se estiver vazia, início e fim apontam para o novo nó
        else:
            self.fim.prox = novo_no            # O último nó passa a apontar para o novo nó
            self.fim = novo_no                 # Atualiza o fim para o novo nó
        self._tamanho += 1                     # Aumenta o tamanho da fila

    # Remove o primeiro elemento da fila
    def desenfileirar(self):
        if self.esta_vazia():                  # Verifica se a fila está vazia
            return "Fila Vazia!"

        valor = self.inicio.valor              # Guarda o valor do primeiro elemento
        self.inicio = self.inicio.prox         # O início passa a ser o próximo nó

        # se a fila ficou vazia, o fim também deve ser nulo:
        if self.inicio is None:                # Se depois de remover não sobrou ninguém
            self.fim = None                    # O fim também vira None

        self._tamanho -= 1                     # Diminui o tamanho da fila
        return valor                           # Retorna o valor removido

    # Retorna o primeiro valor da fila se existir
    def ver_inicio(self):
        if self.inicio.valor is None:
            return "Fila Vazia"
        else:
            return self.inicio.valor

    # Remove todos os nós da fila
    def destruir(self):
        while self.inicio is not None:
            temp = self.inicio              # Guarda o nó atual
            self.inicio = self.inicio.prox  # Move o início para o próximo nó
            temp.prox = None                # Remove a ligação do nó

        self.fim = None                     # Fim também deve ser None
        self._tamanho = 0                   # Zera o tamanho
```

```python
# TESTANDO ----------------------------------------------
minha_fila = Fila()
minha_fila.enfileirar("Senha 01")
minha_fila.enfileirar("Senha 02")
minha_fila.enfileirar("Senha 03")
minha_fila.enfileirar("Senha 04")

print(f"Primeiro da fila: {minha_fila.ver_inicio()}")
print(f"Atendendo: {minha_fila.desenfileirar()}")
print(f"Próximo a ser atendido: {minha_fila.ver_inicio()}")
print(f"Clientes restantes: {minha_fila.__len__()}")
```

# 13.3. Pilha, conversão Infixa para PosFixa

```python
# Algoritmo de Shunting-Yard - Conversão de expressões aritméticas da notação infixa para pós-fixa

# Função que define a prioridade dos operadores
def prioridade(op):
    if op in ('+', '-'):        # Se o operador for + ou -
        return 1                # prioridade menor
    if op in ('*', '/'):        # Se o operador for * ou /
        return 2                # prioridade média
    if op == '^':               # Se for potência
        return 3                # prioridade maior
    return 0                    # Se não for operador, prioridade 0


# Função que converte expressão infixa para pós-fixa
def infixa_posfixa(expressao):
    pilha = []                  # Cria a pilha (lista usada como pilha)
    saida = []                  # Cria a lista de saída (expressão pós-fixa)

    # ignorando espaços em branco:
    for caractere in expressao.replace(" ", ""):   # Percorre cada caractere da expressão
        # 1. se é operando, vai para a saída
        if caractere.isalnum():      # Se for letra ou número
            saida.append(caractere)  # Adiciona direto na saída

        # 2. '(' vai para a pilha
        elif caractere == '(':       # Se for parêntese abrindo
            pilha.append(caractere)  # Empilha o parêntese

        # 3. ')': desempilha tudo até '('
        elif caractere == ')':     # Se for parêntese fechando
            while pilha and pilha[-1] != '(':  # Enquanto não encontrar '('
                saida.append(pilha.pop())      # Desempilha e manda para saída
            pilha.pop()                        # Remove o '(' da pilha

        # 4. se for operador aritmético
        else:
            while (pilha and pilha[-1] != '(' and prioridade(caractere) <= prioridade(pilha[-1])):
                # Enquanto a pilha não estiver vazia, topo não for '('
                # e a prioridade do operador atual for menor ou igual ao topo
                saida.append(pilha.pop())      # Desempilha operador para saída
            pilha.append(caractere)            # Empilha o operador atual

    # 5. Esvazia o restante da pilha
    while pilha:                   # Enquanto ainda tiver operador na pilha
        saida.append(pilha.pop())  # Desempilha tudo para saída

    return "".join(saida)      # Junta os caracteres e retorna a expressão pós-fixa
```

```python
# TESTANDO ---------------------------------------------------
expressoes = [
    "A+B",             # Expressão simples
    "A+B*C",           # Multiplicação tem prioridade
    "(A+B)*C",         # Parênteses primeiro
    "A+B*C-D",         # Vários operadores
    "(A+B)*(C-D)",     # Dois parênteses
    "A+B*(C+D)"        # Parênteses com multiplicação
]

for exp in expressoes:                 # Percorre a lista de expressões
    resultado = infixa_posfixa(exp)    # Converte para pós-fixa
    print(f"Infixa:   {exp}")          # Mostra a expressão original
    print(f"Pos-fixa: {resultado}")    # Mostra a expressão convertida
    print("-" * 30)                    # Linha separadora
```

# 13.4.  Avaliação de expressão em notação pós fixa, como pilha

```python
def calcular(expressao):
    pilha = []                              # Cria uma pilha vazia

    for token in expressao.split():          # Percorre cada elemento separado por espaço
        # Se for número, empilha
        if token.isdigit() or (token[0] == '-' and token[1:].isdigit()):
            pilha.append(float(token))       # Converte para float e empilha
        else:                                # Se for operador, desempilha dois valores
            op2 = pilha.pop()                # Remove o topo (segundo operando)
            op1 = pilha.pop()                # Remove o próximo (primeiro operando)

            if token == '+':                 # Soma
                resultado = op1 + op2
            elif token == '-':               # Subtração
                resultado = op1 - op2
            elif token == '*':               # Multiplicação
                resultado = op1 * op2
            elif token == '/':               # Divisão
                resultado = op1 / op2
            elif token == '^':               # Potência
                resultado = op1 ** op2

            pilha.append(resultado)          # Empilha o resultado

    return pilha[0]                          # Resultado final está no topo
```

```python
# TESTANDO ---------------------------------------------------
expressoes = [
    "2 3 +",            # 2 + 3
    "2 3 5 * +",        # 2 + (3 * 5)
    "2 3 + 5 *",        # (2 + 3) * 5
    "10 2 8 * + 3 -",   # 10 + (2 * 8) - 3
    "5 1 2 + 4 * + 3 -",# 5 + ((1 + 2) * 4) - 3
    "7 2 3 * -",        # 7 - (2 * 3)
    "4 2 / 3 +",        # (4 / 2) + 3
    "2 3 ^",            # 2 ^ 3
    "9 3 / 2 * 5 +",    # ((9 / 3) * 2) + 5
]

for exp in expressoes:                 # Percorre a lista de expressões
    resultado = calcular(exp)
    print(f"Expressão pós-fixa: {exp}")
    print(f"Resultado: {resultado}")
    print("-" * 30)
```
