# 14. Lista Duplamente Encadeada Circular

```python
# Classe que representa cada página (cada nó da lista)
class Pagina:
    def __init__(self, url):          # construtor da classe
        self.url = url                # guarda o endereço da página
        self.proxima = None           # ponteiro para a próxima página
        self.anterior = None          # ponteiro para a página anterior


# Classe que representa o histórico de navegação (lista circular)
class HistoricoDeNavegacao:
    # o navegador começa com uma página inicial (home) -------------------------------------
    def __init__(self, url_inicial):
        nova_pagina = Pagina(url_inicial)     # cria a primeira página
        # como é circular, a página aponta para ela mesma
        nova_pagina.proxima = nova_pagina     # próxima é ela mesma
        nova_pagina.anterior = nova_pagina    # anterior é ela mesma
        self.pagina_atual = nova_pagina       # página atual começa sendo a home
        print(f"Iniciando navegacao em {self.pagina_atual.url}")


    # visitar uma nova página (inserir após a página atual) -------------------------------------
    def visitar_nova_pagina(self, url):
        nova = Pagina(url)                    # cria nova página
        # liga a nova página na lista circular
        nova.anterior = self.pagina_atual         # anterior da nova é a atual
        nova.proxima = self.pagina_atual.proxima  # próxima da nova é a próxima da atual
        # ajusta os ponteiros dos vizinhos
        self.pagina_atual.proxima.anterior = nova  # a próxima agora aponta para nova
        self.pagina_atual.proxima = nova           # atual aponta para nova

        self.pagina_atual = nova                    # nova página vira a página atual
        print(f"Visitando a pagina: {self.pagina_atual.url}")


    # voltar para a página anterior -------------------------------------
    def voltar(self):
        # em lista circular sempre existe anterior
        self.pagina_atual = self.pagina_atual.anterior
        print(f"Voltando para a página: {self.pagina_atual.url}")


    # avançar para a próxima página -------------------------------------
    def avancar(self):
        # em lista circular sempre existe próxima
        self.pagina_atual = self.pagina_atual.proxima
        print(f"Avançando para a página: {self.pagina_atual.url}")
```

```python
# TESTANDO -------------------------------
browser = HistoricoDeNavegacao("google.com")   # página inicial
browser.visitar_nova_pagina("github.com")      # visita github
browser.visitar_nova_pagina("stackoverflow.com")  # visita stackoverflow
browser.visitar_nova_pagina("youtube.com")     # visita youtube

print("\n--- Voltando ---")
browser.voltar()   # youtube -> stackoverflow
browser.voltar()   # stackoverflow -> github

print("\n--- Avançando ---")
browser.avancar()  # github -> stackoverflow
```
