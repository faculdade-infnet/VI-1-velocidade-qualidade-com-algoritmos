# pi - Gregory-Leibiz
- Calcula Quantos termos da série são necessários para obtermos os valores dentro do Array
- Apróximação do número PI

```python
pi = 0 # variavel
sinal = -1 # A cada passo multiplico por -1
i = 0 # Contador de passos
indice = 0 # contador de item do array
L = [3.14 , 3.141, 3.1415, 3.14159] # array com possíveis respostas

while (indice < 4):
  sinal = (-1)*sinal
  pi = pi + 4 * sinal/(2*i + 1)
  i = i + 1

  # round pois é um float com casas decimais talvez infinitas
  # verifica se o numneor encontra = igual ao array
  if (round(pi, indice + 2)) == L[indice]:
    print(L[indice], "encontrado com ", i, "passos.")
    # se for igual ao indice do array aumenta o valor do indice em +1
    indice = indice + 1
```
