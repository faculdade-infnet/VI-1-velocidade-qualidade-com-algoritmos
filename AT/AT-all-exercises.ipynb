# 1 - Busca linear e Busca Binária

```python
import random

# BUSCA LINEAR -------------------------------------------------------------
def busca_linear(arr, target):
    n = len(arr)
    cont = 0                    # conta número de passos executados

    for i in range(n):
        cont += 1
        if arr[i] == target:
            return i, cont      # retorna posição encontrada
    return -1, cont             # retorna se não encontrou nada

# PRÉ-CONDIÇÃO: vetor deve estar ordenado ----------------------------------
# FALHA RÁPIDA: se NÃO estiver ordenado ------------------------------------
def esta_ordenado(arr):
    # percorre o comprimentodo array,
    for i in range(len(arr)-1):
        # caso algum valor nõa esteja ordenado (menor --> maior)
        if arr[i] > arr[i+1]:
            return False
    return True

# BUSCA BINÁRIA ------------------------------------------------------------
def busca_binaria(arr, target):
    # pré-condição
    if not esta_ordenado(arr):
        return -1, 0        # falha rápida

    inicio = 0
    fim = len(arr)-1
    cont = 0                # conta número de passos executados

    while inicio <= fim:
        cont += 1
        meio = (inicio + fim)//2    # divisão inteira

        if arr[meio] == target:
            return meio, cont
        if arr[meio] > target:
            fim = meio - 1
        else:
            inicio = meio + 1

    return -1, cont


# GERAR VETOR --------------------------------------------------------------
def gerar_vetor(n, ordenado=False):
    arr = [random.randint(0, n) for _ in range(n)]
    return sorted(arr) if ordenado else arr
```

```python
# TESTANDO -----------------------------------------------------------------
buscar_valor = 11        # valor a ser buscado
encontrado = False       # verifica se valor foi encontrado

print(f"Nº de comparações ao buscar ({buscar_valor})")
print(f"---------------------------------------")

for n in [100, 1000, 10000, 100000]:
    arr = gerar_vetor(n, ordenado=True) # gera vetor ordenado ou não ordenado

    # faz busca linear e retrona posição e número de comparações
    pos_lin, cont_lin = busca_linear(arr, buscar_valor)
    # faz busca binária e retrona posição e número de comparações
    pos_bin, cont_bin = busca_binaria(arr, buscar_valor)

    # verifica se encontrou o vlaor em alguma busca
    if pos_lin != -1 or pos_bin != -1:
        encontrado = True

    print(f"n={n:<6} | linear={cont_lin:<6} | binaria={cont_bin}")

# mensagem final
print("\nResultado da busca:")
if encontrado:
    print(f"Nº {buscar_valor}, encontrado.")
else:
    print(f"Nº {buscar_valor}, NÃO encontrado.")
```

# 2 - Contagem, bubble_sort, selection_sort, insertion_sort, BST(binary search tree)

```python
import random
import copy
import sys
sys.setrecursionlimit(200000)

# BUBBLE SORT ------------------------------------------------------
def bubble_sort(arr):
    n = len(arr)
    comp = 0        # conta número de comparações
    copias = 0

    # Loop externo, controla o número de passagens do algoritmo
    for i in range(n):
        # Loop interno, compara os elementos vizinhos
        for j in range(0, n - i - 1):
            comp += 1
            if arr[j] > arr[j + 1]:
                # Troca os elementos = 3 cópias
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                copias += 3

    return comp, copias


# SELECTION SORT ------------------------------------------------------
def selection_sort(arr):
    n = len(arr)
    comp = 0        # conta número de comparações
    copias = 0      # conta número de copias

    # percorre a lista para localizar o menor elemento
    for i in range(n):
        min_index = i

        # encontramos o menor elemento na parte não ordenada da lista
        for j in range(i + 1, n):
            comp += 1
            # verifica se encontrou um valor menor
            if arr[j] < arr[min_index]:
                min_index = j

        # trocando o menor elemento com o primeiro elemento da parte não ordenada
        if min_index != i:
            # Troca os elementos = 3 cópias
            arr[i], arr[min_index] = arr[min_index], arr[i]
            copias += 3

    return comp, copias


# INSERTION SORT ------------------------------------------------------
def insertion_sort(arr):
    n = len(arr)
    comp = 0        # conta número de comparações
    copias = 0      # conta número de copias

    # percorre a lista a partir do segundo elemento
    for i in range(1, n):
        chave = arr[i]      # o elemento atual é o escolhido para inserção
        j = i - 1           # inicia a comparação a partir do elemento anterior da lista

        # mover para direita os elementos da sub-lista ordenada que são maiores do que "chave"
        while j >= 0:
            comp += 1
            if arr[j] > chave:
                arr[j + 1] = arr[j]
                copias += 1
                j -= 1
            else:
                break

        # insere chave na posição correta
        arr[j + 1] = chave
        copias += 1

    return comp, copias


# BST - BINARY SEARCH TREE ------------------------------------------------------
# representa um nó da árvore (cada número vira um nó)
class Node:
    def __init__(self, valor):
        self.valor = valor      # valor do nó
        self.esq = None         # ponteiro do filho da esquerda
        self.dir = None         # ponteiro do filho da direita


class BST:
    def __init__(self):
        self.raiz = None
        self.comp = 0       # conta número de comparações
        self.chamadas = 0   # conta número de chmadas

    # função principal de inserção
    def inserir(self, valor):
        # se árvore= vazia, cria a raiz, senão, chama função auxiliar
        if self.raiz is None:
            self.raiz = Node(valor)
        else:
            self._inserir(self.raiz, valor)

    # função recursiva para inserir respeitando a regra da BST
    def _inserir(self, node, valor):
        self.comp += 1
        # se o valor for menor, vai para esquerda
        if valor < node.valor:
            if node.esq is None:
                node.esq = Node(valor)      # insere se estiver vazio
            else:
                self._inserir(node.esq, valor)  # continua descendo
        else:           # se for maior ou igual, vai para direita
            if node.dir is None:
                node.dir = Node(valor)
            else:
                self._inserir(node.dir, valor)

    # percorre a árvore (esquerda -> direita)
    def in_ordem(self, node):
        if node:
            self.chamadas += 1          # CONT visita a um nó
            self.in_ordem(node.esq)     # percorre á esquerdo
            self.in_ordem(node.dir)     # percorre á direita

# cria a árvore, insere os valores, percorre a estrutura e ordena
def bst_sort(arr):
    bst = BST()  # cria uma árvore vazia

    # insere todos os elementos do vetor na árvore
    for v in arr:
        bst.inserir(v)

    bst.in_ordem(bst.raiz)      # percorre a árvore ordenando
    return bst.comp, bst.chamadas


# GERAR VETORES ---------------------------------------------------
def gerar_vetor(n, tipo):
    arr = list(range(n))

    if tipo == "ordenado":
        return arr
    elif tipo == "reverso":
        return arr[::-1]
    # quase ordenado
    elif tipo == "quase":
        arr = arr[:]
        for _ in range(int(n * 0.05)):  # 5% bagunçado
            i = random.randint(0, n - 1)
            j = random.randint(0, n - 1)
            arr[i], arr[j] = arr[j], arr[i]
        return arr
    elif tipo == "aleatorio":
        random.shuffle(arr)
        return arr
```

```python
# TESTANDO ------------------------------
# tamanhos = [1000, 10000, 25000, 50000, 100000]
vetores = [1000, 10000, 25000]  # reduz
# tipos de vetores
vet_tipos = ["ordenado", "reverso", "quase", "aleatorio"]

for n in vetores:
    print(f"\n===== TAMANHO: {n} =====")

    # percorre cada tipo de vetor
    for tipo in vet_tipos:
        print(f"\n--- Tipo: {tipo} ---")

        # gera vetor conforme tipo
        base = gerar_vetor(n, tipo)

        algoritmos = [
            ("Bubble", bubble_sort),
            ("Selection", selection_sort),
            ("Insertion", insertion_sort),
        ]

        # testa os algoritmos de ordenação
        for nome, func in algoritmos:
            arr = copy.copy(base)       # cria uma cópia
            comp, copias = func(arr)    # executa o algoritmo e poega informações
            print(f"{nome}: Comparações={comp}, Cópias={copias}")

        arr = copy.copy(base)
        comp, chamadas = bst_sort(arr)
        print(f"BST: Comparações={comp}, Chamadas={chamadas}")
```

# 3 - Remoção de Duplicados + k menores

```python
import random
import time
import sys
sys.setrecursionlimit(200000)

# HASH TABLE -------------------------------
class HashTable:
    def __init__(self, size):
        self.size = size
        self.tabela = [[] for _ in range(size)]

    def _funcao_hash(self, chave):
        return hash(chave) % self.size

    # insere um novo valor caso ele não exista, se existir atualiza
    def inserir(self, chave, valor):
        index = self._funcao_hash(chave)
        bucket = self.tabela [index]

        # verificar se a chave já existe e atualização do valor
        for i, (chave_existente, _) in enumerate (bucket):
            if chave_existente == chave:
                bucket[i] = (chave, valor)
                return
        # se a chave não existe, adicionar o novo par chave-valor no bucket
        bucket.append((chave, valor))

    # busca um valor dentro do bucket/tabela
    def buscar(self, chave):
        # calcula o índice
        index= self._funcao_hash (chave)
        bucket = self.tabela [index]
        # procurar pela chave e retornar o valor correspondente
        for chave_existente, valor in bucket:
            if chave_existente == chave:
                return valor

        return None # chave não encontrada!

# REMOVER DUPLICADOS O(n) -------------------------------
def remover_duplicados(arr):
    hash_table = HashTable(len(arr))
    resultado = []

    for item in arr:
        if hash_table.buscar(item) is None:
            hash_table.inserir(item, 1)
            resultado.append(item)

    return resultado

# VERSÃO A (ORDENAÇÃO COMPLETA) -------------------------------
# QuickSort usando o mesmo partition
def quicksort(arr, left, right):
    # Verifica se ainda há mais de um elemento no intervalo (parada da recursão)
    if left < right:
        # função que organiza os elementos e retorna a posição final do pivo
        pivot_index = partition(arr, left, right)

        quicksort(arr, left, pivot_index - 1)   # recursivamente á esquerda do pivo
        quicksort(arr, pivot_index + 1, right)  # recursivamente á direita do pivo


# VERSÃO A (ORDENAÇÃO COMPLETA)
def k_smallest_sort(arr, k):
    arr_copy = arr.copy()       # copia do array original
    quicksort(arr_copy, 0, len(arr_copy) - 1)   # rdena completamente o array
    # k menores (primeiros após ordenação)
    return arr_copy[:k]


# QUICKSELECT (TP3) ----------------------------------------------------
# ajusta o array colocando o pivo na posição correta, conm menores à esquerda e maiores à direita
def partition(arr, left, right):
    pivot = arr[right]      # define o pivo como o último elemento do intervalo
    i = left                # indice par os menores que o pivo

    # percorre todos os elementos (exceto pivo)
    for j in range(left, right):
        # se item atual é menor ou igual
        if arr[j] <= pivot:
            arr[i], arr[j] = arr[j], arr[i] # troca atual com o elemento na posição i
            i += 1       # avança índice dos menores

     # coloca o pivo na posição correta (entre menores e maiores)
    arr[i], arr[right] = arr[right], arr[i]
    return i    # posição final do pivo


def quickselect(arr, left, right, k):
    # continua enquanto houver mais de um elemento no intervalo
    if left < right:
        # divide o array e pega a posição do pivo
        pivot_index = partition(arr, left, right)

        # se pivo=k, encontramos
        if pivot_index == k:
            return
        # se pivo maior, ficou à direita de k, busca a esquerda
        elif pivot_index > k:
            quickselect(arr, left, pivot_index - 1, k)
        # Se o pivo menor, ficou à esquerda de k, busca a direita
        else:
            quickselect(arr, pivot_index + 1, right, k)

# retorna os k menores do array, sem ordenar tudo
def k_smallest_quickselect(arr, k):
    arr_copy = arr.copy()       # cria copia
    quickselect(arr_copy, 0, len(arr_copy) - 1, k - 1)  # chamada
    return arr_copy[:k]
```

```python
# TESTANDO ------------------------------
vatores = [1000, 10000, 25000, 50000, 100000]
k = 10

for n in vatores:
    print(f"\n===== TAMANHO: {n} =====")

    # Geração do array aleatório
    inicio_total = time.time()
    arr = [random.randint(0, n // 2) for _ in range(n)]

    # Remoção de duplicados
    inicio = time.time()
    arr_sem_dup = remover_duplicados(arr)
    tempo_dup = time.time() - inicio

    # Versão A (ordenação completa)
    inicio = time.time()
    res_a = k_smallest_sort(arr_sem_dup, k)
    tempo_a = time.time() - inicio

    # Versão B (QuickSelect)
    inicio = time.time()
    res_b = k_smallest_quickselect(arr_sem_dup, k)
    tempo_b = time.time() - inicio

    # Tempo total do processo (geração + remoção + k menores)
    tempo_total = time.time() - inicio_total

    print(f"Remoção duplicados: {tempo_dup:.5f}s | tamanho final: {len(arr_sem_dup)}")
    print(f"Sort (A):           {tempo_a:.5f}s")
    print(f"QuickSelect (B):    {tempo_b:.5f}s")
    print(f"Tempo total:        {tempo_total:.5f}s")

    # Validação
    print(f"Iguais? {sorted(res_a) == sorted(res_b)}")
```

# 4 - HashTableChained

```python
# Classe para o par chave-valor --------------------------------------------------
class Node:
    def __init__(self, key, value):
        self.key = key        # chave do par
        self.value = value    # valor associado
        self.next = None      # ponteiro para o próximo nó (encadeamento)


# Hash Table com encadeamento --------------------------------------------------
class HashTableChained:
    def __init__(self, capacity=4, load_factor=0.75):
        self.capacity = capacity               # número inicial de buckets
        self.size = 0                          # quantidade atual de elementos
        self.buckets = [None] * capacity       # lista de buckets (listas encadeadas)
        self.load_factor = load_factor         # limiar de carga para rehash
        self.comparacoes = 0                   # contador de comparações para teste

    # Função hash simples baseada no hash nativo do Python -------------------------------------------
    def _hash(self, key):
        return hash(key) % self.capacity

    # Rehash automático aumentando a capacidade (dobrando) -------------------------------------------
    def _rehash(self):
        old_buckets = self.buckets            # guarda os buckets antigos
        self.capacity *= 2                     # dobra a capacidade
        self.buckets = [None] * self.capacity # cria nova tabela vazia
        self.size = 0                          # reset do tamanho
        for head in old_buckets:               # percorre cada bucket antigo
            node = head
            while node:
                self.put(node.key, node.value) # insere na nova tabela
                node = node.next

    # Inserir ou atualizar um par (key, value) --------------------------------------------------
    def put(self, key, value):
        if self.size / self.capacity >= self.load_factor:
            self._rehash()                    # rehash se fator de carga exceder

        index = self._hash(key)
        node = self.buckets[index]
        prev = None

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1             # contar comparação
            # se encontrar atualiza o valor
            if node.key == key:
                node.value = value            # atualiza valor existente
                return
            prev = node
            node = node.next

        # se não eocntrar cira um novo nó no final da árvore
        new_node = Node(key, value)           # cria novo nó
        if prev:
            prev.next = new_node              # adiciona ao final da lista encadeada
        else:
            self.buckets[index] = new_node    # bucket estava vazio
        self.size += 1

    # Recuperar valor pela chave --------------------------------------------------
    def get(self, key):
        index = self._hash(key)
        node = self.buckets[index]

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1
            # se a chave do nó atual = chave procurada
            if node.key == key:
                return node.value       # rewtorna o nó atual
            node = node.next            # vai pra o próximo nó
        return None                     # chave não encontrada

    # Remover par pela chave -----------------------------------------------------
    def delete(self, key):
        index = self._hash(key)
        node = self.buckets[index]
        prev = None

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1
            # se a chave do nó atual = chave procurada
            if node.key == key:
                # se não é primeiro nó da lista
                if prev:
                    prev.next = node.next
                else:   # se é o primeiro nó do bucket
                    self.buckets[index] = node.next # bucket para o próximo nó
                self.size -= 1
                return True                   # removeu com sucesso
            prev = node
            node = node.next
        return False                          # chave não encontrada

    # Retorna o número de elementos na tabela --------------------------------------------------
    def __len__(self):
        return self.size
```

```python
# TESTANDO ------------------------------
# Criar tabela pequena para forçar rehash
ht = HashTableChained(capacity=2, load_factor=0.7)
comparacoes_antes = ht.comparacoes

# Inserir elementos
for i in range(10):
    ht.put(f"chave{i}", i)

# Comparações para inserção
comparacoes_insercao = ht.comparacoes
print(f"Comparações na inserção de 10 elementos: {comparacoes_insercao}")

# Comparações para busca de todas as chaves
ht.comparacoes
for i in range(10):
    ht.get(f"chave{i}")
comparacoes_busca = ht.comparacoes
print(f"Comparações na busca de 10 elementos: {comparacoes_busca}")

# Fator de carga atual
fator_carga = ht.size / ht.capacity
print(f"Fator de carga atual: {fator_carga:.2f}")

# Relação custo esperado
print("Custo esperado de busca e inserção é O(1 + α), ou seja:")
print(f"O(1 + {fator_carga:.2f}) ≈ O({1 + fator_carga:.2f}) comparações por operação")
```

# 5 - Árvore + BFS(Busca em Largura)/DFS(Busca em Profundidade)

```python
import random
from collections import deque  # fila pra BFS

# NÓ DA ÁRVORE -------------------------------
class Node:
    def __init__(self, valor):
        self.valor = valor
        self.esq = None
        self.dir = None


# ÁRVORE BINÁRIA DE BUSCA (BST) -------------------------------
class BST:
    def __init__(self):
        self.raiz = None

    def inserir(self, valor):
        self.raiz = self._inserir_rec(self.raiz, valor)

    def _inserir_rec(self, nodo, valor):
        if nodo is None:  # se não tem nó, cria um novo
            return Node(valor)
        if valor < nodo.valor:  # menor vai pra esquerda
            nodo.esq = self._inserir_rec(nodo.esq, valor)
        else:  # maior ou igual vai pra direita
            nodo.dir = self._inserir_rec(nodo.dir, valor)
        return nodo


# DFS recursivo (tipo Questão 4) -------------------------------
def dfs_recursivo(nodo, nivel=0):
    if nodo is None:
        return []
    resultado = []
    resultado.append("  " * nivel + str(nodo.valor))  # tipo mostrar a arvore com espaços
    resultado += dfs_recursivo(nodo.esq, nivel + 1)   # vai pra esquerda primeiro
    resultado += dfs_recursivo(nodo.dir, nivel + 1)   # depois pra direita
    return resultado


# DFS com pilha (profundidade) -------------------------------
def dfs_pilha(raiz):
    if raiz is None:
        return []
    resultado = []
    pilha = [raiz]
    while pilha:
        nodo = pilha.pop()  # pega o ultimo
        resultado.append(nodo.valor)
        if nodo.dir:  # coloca direita primeiro pra esquerda sair primeiro
            pilha.append(nodo.dir)
        if nodo.esq:
            pilha.append(nodo.esq)
    return resultado


# BFS com fila e contagem de comparações -------------------------------
def bfs_fila(raiz):
    if raiz is None:
        return [], 0
    resultado = []
    fila = deque()
    fila.append(raiz)  # coloca raiz na fila
    comparacoes = 0
    while fila:
        nodo = fila.popleft()      # tira da frente da fila
        resultado.append(nodo.valor)
        comparacoes += 1           # conta como se fosse comparacao kk
        if nodo.esq:
            fila.append(nodo.esq)  # adiciona filho esquerdo
        if nodo.dir:
            fila.append(nodo.dir)  # adiciona filho direito
    return resultado, comparacoes
```

```python
# TESTANDO ------------------------------
bst = BST()
elementos = random.sample(range(1, 101), 10)  # pega 10 numeros aleatorios
for val in elementos:
    bst.inserir(val)  # insere na arvore

print("Elementos inseridos:", elementos)
print("\nDFS recursivo (com indentacao):")
print("\n".join(dfs_recursivo(bst.raiz)))  # mostra arvore hierarquica

print("\nDFS com pilha (profundidade):", dfs_pilha(bst.raiz))

bfs_result, bfs_comps = bfs_fila(bst.raiz)
print("BFS com fila (largura):", bfs_result)
print("Numero de comparacoes BFS:", bfs_comps)
```

# 6 - Lista Encadeada Simples

```python
# Nó da lista (cada caixinha) -------------------------------
class No:
    def __init__(self, dado):
        self.dado = dado      # valor guardado na caixinha
        self.prox = None      # ponteiro pro próximo nó (começa vazio)


# Lista Encadeada Simples  -------------------------------
class ListaEncadeadaSimples:
    def __init__(self):
        self.inicio = None    # começo da lista
        self.tamanho = 0      # pra saber quantos elementos tem

    # Csuto Provavel: O(1), bem rapido pq só muda ponteiro
    # inserir no começo da lista -------------------------------
    def inserir_no_inicio(self, dado):
        novo_no = No(dado)
        novo_no.prox = self.inicio   # novo nó aponta pro que era primeiro
        self.inicio = novo_no        # ele vira o primeiro
        self.tamanho += 1


    # Custo Provavel: O(n), tem que percorrer tudo se for grande
    # inserir no final da lista
    def inserir_no_fim(self, dado):
        novo_no = No(dado)
        if not self.inicio:          # se lista ta vazia
            self.inicio = novo_no
        else:
            atual = self.inicio
            while atual.prox:         # vai andando até o último nó
                atual = atual.prox
            atual.prox = novo_no     # liga o último nó ao novo
        self.tamanho += 1


    # Custo porvavel: O(n), porque precisa andar até a posição
    # inserir numa posição específica -------------------------------
    def inserir_na_posicao(self, indice, dado):
        if indice < 0 or indice > self.tamanho:
            print(f"Erro: índice {indice} inválido")
            return

        if indice == 0:                # se for primeiro, usa inserir_no_inicio
            self.inserir_no_inicio(dado)
            return

        novo_no = No(dado)
        atual = self.inicio
        for _ in range(indice-1):      # vai até o nó antes da posição
            atual = atual.prox
        novo_no.prox = atual.prox      # nó novo aponta pro próximo
        atual.prox = novo_no           # anterior aponta pro nó novo
        self.tamanho += 1


    # Custo provavel: O(n), tem que olhar quase tudo se for no final
    # procurar valor -------------------------------
    def buscar(self, valor):
        atual = self.inicio
        pos = 0
        while atual:
            if atual.dado == valor:
                print(f"Valor {valor} encontrado na posição {pos}")
                return pos
            atual = atual.prox
            pos += 1
        print(f"Valor {valor} não encontrado")
        return -1


    # Custo provavel: O(n), tem que percorrer pra achar
    # remover por valor -------------------------------
    def remover(self, valor):
        atual = self.inicio
        anterior = None

        while atual and atual.dado != valor:
            anterior = atual
            atual = atual.prox

        if not atual:
            print(f"Valor {valor} não encontrado")
            return

        if anterior is None:          # remover primeiro nó
            self.inicio = atual.prox
        else:
            anterior.prox = atual.prox

        self.tamanho -= 1
        print(f"Valor {valor} removido")


    # Custo provavel: O(n), precisa andar até o nó certo
    # remover por posição -------------------------------
    def remover_na_posicao(self, indice):
        if indice < 0 or indice >= self.tamanho:
            print(f"Erro: índice {indice} inválido")
            return

        atual = self.inicio
        anterior = None

        for _ in range(indice):        # vai até a posição
            anterior = atual
            atual = atual.prox

        if anterior is None:           # se for primeiro
            self.inicio = atual.prox
        else:
            anterior.prox = atual.prox

        self.tamanho -= 1
        print(f"Nó na posição {indice} removido")


    # Custo provavel: O(1), super rapido
    # retorna tamanho da lista -------------------------------
    def __len__(self):
        return self.tamanho           # só devolver o contador


    # Custo provavel: O(n), tem que percorrer todos os nós
    # mostrar a lista como string -------------------------------
    def __str__(self):
        elementos = []
        atual = self.inicio
        while atual:
            elementos.append(str(atual.dado))
            atual = atual.prox
        return " -> ".join(elementos) + " -> $"
```

```python
# TESTANDO ------------------------------
lista = ListaEncadeadaSimples()

lista.inserir_no_inicio(10)       # insere o item 10 no inicio
lista.inserir_no_fim(20)          # insere o item 20 no final
lista.inserir_na_posicao(1, 15)   # insere o item 15 na posicao 1
print(lista)

lista.buscar(15)                   # procura o elemento 15
lista.remover(20)                  # remove o elemento 20
lista.remover_na_posicao(0)        # remove o primeiro
print(lista)
print("Tamanho:", len(lista))
```

# 7. Lista duplamente encadeada + Deque

```python
# Nodo da lista duplamente encadeada --------------------------
class Nodo:
    def __init__(self, valor):
        self.valor = valor         # guarda o valor do nodo
        self.anterior = None       # ponteiro pro nodo anterior
        self.proximo = None        # ponteiro pro próximo nodo


# Lista duplamente encadeada --------------------------
class ListaDuplamenteEncadeada:
    def __init__(self):
        self.cabeca = None         # começo da lista
        self.cauda = None          # fim da lista
        self.tamanho = 0           # contador de elementos (igual na pilha TP2)

    def esta_vazia(self):
        return self.tamanho == 0   # True se não tem nada

    # inserir no começo da lista -------------------------------
    def inserir_inicio(self, valor):
        novo = Nodo(valor)
        if self.esta_vazia():       # se lista vazia, cabeça e cauda viram o novo nodo
            self.cabeca = self.cauda = novo
        else:
            novo.proximo = self.cabeca
            self.cabeca.anterior = novo
            self.cabeca = novo
        self.tamanho += 1

    # inserir no fim da lista -------------------------------
    def inserir_fim(self, valor):
        novo = Nodo(valor)
        # Se a lista estiver vazia, então ele vira a cabeça e a cauda ao mesmo tempo
        if self.esta_vazia():
            self.cabeca = self.cauda = novo
        else:       # O novo nodo aponta pro antigo último nodo (anterior)
            novo.anterior = self.cauda
            self.cauda.proximo = novo
            self.cauda = novo
        self.tamanho += 1

    # remover do começo da lista -------------------------------
    def remover_inicio(self):
        if self.esta_vazia():
            print("Lista vazia! Nada pra remover.")
            return None

        removido = self.cabeca  # Guarda o nodo que vamos remover
        if self.cabeca == self.cauda:   # só tinha um elemento
            self.cabeca = self.cauda = None
        else:
            self.cabeca = self.cabeca.proximo
            self.cabeca.anterior = None
        self.tamanho -= 1
        return removido.valor

    # remover do fim da lista -------------------------------
    def remover_fim(self):
        if self.esta_vazia():
            print("Lista vazia! Nada pra remover.")
            return None

        removido = self.cauda       # Guarda o nodo que vamos remover

        # Se são iguais é pq só tinha um elemento
        if self.cabeca == self.cauda:
            self.cabeca = self.cauda = None
        else:           # Senao, então tem mais elementos, a cabeça agora vira o próximo nodo
            self.cauda = self.cauda.anterior
            self.cauda.proximo = None

        self.tamanho -= 1       # -1, porque removemos um elemento
        return removido.valor

    # imprimir a lista -------------------------------
    def imprimir(self):
        atual = self.cabeca
        valores = []
        while atual:
            valores.append(atual.valor)
            atual = atual.proximo
        print("Lista: ", valores)


    # Verificação das invariantes estruturais  --------------------------
    def verificar_invariantes(self):
        # Se a lista estiver vazia, a cabeça e a cauda devem ser None
        if self.esta_vazia():
            if self.cabeca is not None or self.cauda is not None:
                print("Erro: lista vazia mas cabeca ou cauda não são None!")
            return True  # tudo certo, lista vazia

        # checando cabeça e cauda
        if self.cabeca.anterior is not None:
            print("Erro: cabeca.anterior não é None")  # cabeça não deve ter anterior
        if self.cauda.proximo is not None:
            print("Erro: cauda.proximo não é None")    # cauda não deve ter próximo

        atual = self.cabeca
        contador = 0
        while atual:        # percorrer a lista e checar ponteiros
            contador += 1
            # checar se proximo nodo aponta de volta pro atual
            if atual.proximo and atual.proximo.anterior != atual:
                print(f"Erro: nodo {atual.valor}.proximo.anterior != nodo atual")
            # checar se nodo anterior aponta de volta pro atual
            if atual.anterior and atual.anterior.proximo != atual:
                print(f"Erro: nodo {atual.valor}.anterior.proximo != nodo atual")
            atual = atual.proximo

        # checar se o tamanho registrado bate com a contagem real
        if contador != self.tamanho:
            print(f"Erro: tamanho armazenado ({self.tamanho}) != contagem real ({contador})")

        # retorno True se tudo bateu certinho, False se deu algum problema
        return contador == self.tamanho


# Deque usando a Lista Duplamente encadeada --------------------------
class Deque:
    def __init__(self):
        self.lista = ListaDuplamenteEncadeada()  # usa a lista dupla

    def inserir_esquerda(self, valor):      # inserir à esquerda (começo da lista)
        self.lista.inserir_inicio(valor)

    def inserir_direita(self, valor):       # inserir à direita (fim da lista)
        self.lista.inserir_fim(valor)

    def remover_esquerda(self):             # remover da esquerda (começo da lista)
        return self.lista.remover_inicio()

    def remover_direita(self):              # remover da direita (fim da lista)
        return self.lista.remover_fim()

    def espiar_esquerda(self):              # espiar o elemento da esquerda
        if self.lista.esta_vazia():
            print("Deque vazio!")
            return None
        return self.lista.cabeca.valor

    def espiar_direita(self):               # espiar o elemento da direita
        if self.lista.esta_vazia():
            print("Deque vazio!")
            return None
        return self.lista.cauda.valor

    # imprimir deque
    def imprimir(self):
        self.lista.imprimir()

    # método auxiliar para verificar invariantes
    def checar_invariantes(self):
        print("Verificando invariantes...")  # comentário estilo aluno
        ok = self.lista.verificar_invariantes()
        if ok:
            print("Tudo certo! Ponteiros consistentes e tamanho correto.")
        else:
            print("Erro nas invariantes!")  # se algo estiver errado nos ponteiros ou tamanho
```

```python
# TESTANTO ----------------------------------------------
deque = Deque()

# inserções alternadas
deque.inserir_esquerda(10)
deque.inserir_direita(20)
deque.inserir_esquerda(5)
deque.inserir_direita(25)
print("Depois das inserções:")
deque.imprimir()
deque.checar_invariantes()   # verifica ponteiros + tamanho

# espiar os dois lados
print("Espiar esquerda:", deque.espiar_esquerda())
print("Espiar direita:", deque.espiar_direita())

# remoções alternadas
print("Remover esquerda:", deque.remover_esquerda())
print("Remover direita:", deque.remover_direita())
print("Depois das remoções:")
deque.imprimir()
deque.checar_invariantes()

# mais operações pra testar invariantes
deque.inserir_direita(30)
deque.inserir_esquerda(2)
deque.inserir_direita(40)
print("Inserções extras:")
deque.imprimir()
deque.checar_invariantes()  # invariantes estruturais verificadas

# verificar tamanho final
print("Tamanho final do deque:", deque.lista.tamanho)
```

# 8. Mochila Knapsack recursivo + Programação dinâmica

```python
# Classe para o par chave-valor --------------------------------------------------
class Node:
    def __init__(self, key, value):
        self.key = key        # chave do par
        self.value = value    # valor associado
        self.next = None      # ponteiro para o próximo nó (encadeamento)


# Hash Table com encadeamento --------------------------------------------------
class HashTableChained:
    def __init__(self, capacity=4, load_factor=0.75):
        self.capacity = capacity               # número inicial de buckets
        self.size = 0                          # quantidade atual de elementos
        self.buckets = [None] * capacity       # lista de buckets (listas encadeadas)
        self.load_factor = load_factor         # limiar de carga para rehash
        self.comparacoes = 0                   # contador de comparações para teste

    # Função hash simples baseada no hash nativo do Python -------------------------------------------
    def _hash(self, key):
        return hash(key) % self.capacity

    # Rehash automático aumentando a capacidade (dobrando) -------------------------------------------
    def _rehash(self):
        old_buckets = self.buckets            # guarda os buckets antigos
        self.capacity *= 2                     # dobra a capacidade
        self.buckets = [None] * self.capacity # cria nova tabela vazia
        self.size = 0                          # reset do tamanho
        for head in old_buckets:               # percorre cada bucket antigo
            node = head
            while node:
                self.put(node.key, node.value) # insere na nova tabela
                node = node.next

    # Inserir ou atualizar um par (key, value) --------------------------------------------------
    def put(self, key, value):
        if self.size / self.capacity >= self.load_factor:
            self._rehash()                    # rehash se fator de carga exceder

        index = self._hash(key)
        node = self.buckets[index]
        prev = None

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1             # contar comparação
            # se encontrar atualiza o valor
            if node.key == key:
                node.value = value            # atualiza valor existente
                return
            prev = node
            node = node.next

        # se não eocntrar cira um novo nó no final da árvore
        new_node = Node(key, value)           # cria novo nó
        if prev:
            prev.next = new_node              # adiciona ao final da lista encadeada
        else:
            self.buckets[index] = new_node    # bucket estava vazio
        self.size += 1

    # Recuperar valor pela chave --------------------------------------------------
    def get(self, key):
        index = self._hash(key)
        node = self.buckets[index]

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1
            # se a chave do nó atual = chave procurada
            if node.key == key:
                return node.value       # rewtorna o nó atual
            node = node.next            # vai pra o próximo nó
        return None                     # chave não encontrada

    # Remover par pela chave -----------------------------------------------------
    def delete(self, key):
        index = self._hash(key)
        node = self.buckets[index]
        prev = None

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1
            # se a chave do nó atual = chave procurada
            if node.key == key:
                # se não é primeiro nó da lista
                if prev:
                    prev.next = node.next
                else:   # se é o primeiro nó do bucket
                    self.buckets[index] = node.next # bucket para o próximo nó
                self.size -= 1
                return True                   # removeu com sucesso
            prev = node
            node = node.next
        return False                          # chave não encontrada

    # Retorna o número de elementos na tabela --------------------------------------------------
    def __len__(self):
        return self.size


# (a) RECURSIVO PURO ------------------------------------------------------------
def knapsack_rec(valores, pesos, capacidade, i=0, chamadas=None, subproblemas=None):

    # mesmo padrão usado no fibonacci (Q8)
    if chamadas is None:
        chamadas = {"total": 0}

    # guarda estados diferentes (subproblemas)
    if subproblemas is None:
        subproblemas = set()

    chamadas["total"] += 1  # conta chamada

    # cada estado é definido por (i, capacidade)
    subproblemas.add((i, capacidade))

    # CASO BASE (acabou itens ou não cabe mais nada)
    if i >= len(valores) or capacidade <= 0:
        return 0, chamadas["total"], len(subproblemas)

    # se o peso não cabe, pula (igual lógica da Q6 - não escolher)
    if pesos[i] > capacidade:
        return knapsack_rec(valores, pesos, capacidade, i + 1, chamadas, subproblemas)

    # 1: escolher o item (igual "adicionar" na Q6)
    pegar, _, _ = knapsack_rec(
        valores, pesos, capacidade - pesos[i], i + 1, chamadas, subproblemas
    )
    pegar += valores[i]

    # 2: não escolher o item (igual Q6)
    nao_pegar, _, _ = knapsack_rec(
        valores, pesos, capacidade, i + 1, chamadas, subproblemas
    )

    # retorna o melhor valor
    return max(pegar, nao_pegar), chamadas["total"], len(subproblemas)


# (b) MEMOIZATION COM HASH TABLE ------------------------------------------------------------
def knapsack_memo(valores, pesos, capacidade, i=0, memo=None, chamadas=None):

    # usa a HashTableChained da Q4 no lugar de dict
    if memo is None:
        memo = HashTableChained()

    # mesmo padrão de contador da Q8
    if chamadas is None:
        chamadas = {"total": 0}

    chamadas["total"] += 1

    # chave do estado (estado mínimo suficiente) (mesma ideia de subproblemas do recursivo)
    chave = f"{i}-{capacidade}"

    # tenta pegar da hash
    valor_guardado = memo.get(chave)
    if valor_guardado is not None:
        return valor_guardado, chamadas["total"]

    # CASO BASE
    if i >= len(valores) or capacidade <= 0:
        memo.put(chave, 0)
        return 0, chamadas["total"]

    # se não cabe, pula (igual recursivo)
    if pesos[i] > capacidade:
        valor, _ = knapsack_memo(valores, pesos, capacidade, i + 1, memo, chamadas)
        memo.put(chave, valor)
        return valor, chamadas["total"]

    # 1: pegar item
    pegar, _ = knapsack_memo(
        valores, pesos, capacidade - pesos[i], i + 1, memo, chamadas
    )
    pegar += valores[i]

    # 2: não pegar
    nao_pegar, _ = knapsack_memo(
        valores, pesos, capacidade, i + 1, memo, chamadas
    )

    melhor = max(pegar, nao_pegar)

    # salva resultado na hash (memoization)
    memo.put(chave, melhor)

    return melhor, chamadas["total"]
```

```python
# TESTANDO (mínimo 16 itens)  ------------------------------------------------------------
valores = [10, 40, 30, 50, 35, 40, 30, 45, 25, 20, 15, 60, 55, 10, 5, 70]
pesos   = [5,  4,  6,  3,  7,  2,  5,  6,  3,  2,  1,  8,  7,  2, 1,  9]

capacidade = 20

print("RESULTADO MOCHILA 0/1")
print("-----------------------------------")

# versão recursiva pura
valor_rec, chamadas_rec, sub_rec = knapsack_rec(valores, pesos, capacidade)

print("Recursivo puro:")
print("Valor máximo:", valor_rec)
print("Chamadas:", chamadas_rec)
print("Subproblemas diferentes:", sub_rec)

# versão com memoization (hash)
valor_memo, chamadas_memo = knapsack_memo(valores, pesos, capacidade)

print("\nCom memoization (HashTable):")
print("Valor máximo:", valor_memo)
print("Chamadas:", chamadas_memo)
```
