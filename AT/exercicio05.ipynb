# 5 - Árvore + BFS(Busca em Largura)/DFS(Busca em Profundidade)

```python
import random
from collections import deque  # deque para fila (BFS)

# NÓ DA ÁRVORE BINÁRIA -------------------------------
class Node:
    def __init__(self, valor):
        self.valor = valor      # valor armazenado no nó
        self.esq = None         # ponteiro para filho esquerdo
        self.dir = None         # ponteiro para filho direito


# ÁRVORE BINÁRIA DE BUSCA (BST) -------------------------------
class BST:
    def __init__(self):
        self.raiz = None        # nó raiz inicial da árvore

    # Inserir um valor na árvore
    def inserir(self, valor):
        self.raiz = self._inserir_rec(self.raiz, valor)

    # Inserção recursiva ------------------------------------
    def _inserir_rec(self, nodo, valor):
        if nodo is None:  # se não existe nó, cria um novo
            return Node(valor)
        if valor < nodo.valor:  # valor menor vai para esquerda
            nodo.esq = self._inserir_rec(nodo.esq, valor)
        else:  # valor maior ou igual vai para direita
            nodo.dir = self._inserir_rec(nodo.dir, valor)
        return nodo


# DFS recursivo com indentação (pré-ordem) -------------------------------
def dfs_recursivo(nodo, nivel=0):
    if nodo is None:
        return []  # nada a retornar
    resultado = []
    # adiciona valor do nó com indentação conforme nível
    resultado.append("  " * nivel + str(nodo.valor))
    # percorre filho esquerdo primeiro
    resultado += dfs_recursivo(nodo.esq, nivel + 1)
    # percorre filho direito
    resultado += dfs_recursivo(nodo.dir, nivel + 1)
    return resultado


# DFS usando pilha (profundidade) -------------------------------
def dfs_pilha(raiz):
    if raiz is None:
        return []
    resultado = []
    pilha = [raiz]  # inicializa pilha com a raiz
    while pilha:
        nodo = pilha.pop()         # remove topo da pilha
        resultado.append(nodo.valor)
        # adiciona filho direito primeiro, assim o esquerdo será processado primeiro
        if nodo.dir:
            pilha.append(nodo.dir)
        if nodo.esq:
            pilha.append(nodo.esq)
    return resultado


# BFS usando fila (largura) e contagem de comparações -------------------------------
def bfs_fila(raiz):
    if raiz is None:
        return [], 0
    resultado = []
    fila = deque()
    fila.append(raiz)  # coloca raiz na fila
    comparacoes = 0
    while fila:
        nodo = fila.popleft()       # remove da frente da fila
        resultado.append(nodo.valor)
        comparacoes += 1            # simula comparação/processamento do nó
        # adiciona filhos na fila
        if nodo.esq:
            fila.append(nodo.esq)
        if nodo.dir:
            fila.append(nodo.dir)
    return resultado, comparacoes
```

```python
# TESTANDO ------------------------------
bst = BST()
# Gerar 10 elementos aleatórios únicos entre 1 e 100
elementos = random.sample(range(1, 101), 10)

# Inserir elementos na árvore
for val in elementos:
    bst.inserir(val)

print("Elementos inseridos na BST:", elementos)

# DFS recursivo para mostrar a árvore hierárquica
print("\nDFS recursivo (pré-ordem com indentação):")
print("\n".join(dfs_recursivo(bst.raiz)))

# DFS com pilha (profundidade)
dfs_profundidade = dfs_pilha(bst.raiz)
print("\nDFS usando pilha (profundidade):", dfs_profundidade)

# BFS com fila (largura)
bfs_result, bfs_comps = bfs_fila(bst.raiz)
print("BFS usando fila (largura):", bfs_result)
print("Número de comparações/processamento BFS:", bfs_comps)
```
