# 4 - HashTableChained

```python
# Classe para o par chave-valor --------------------------------------------------
class Node:
    def __init__(self, key, value):
        self.key = key        # chave do par
        self.value = value    # valor associado
        self.next = None      # ponteiro para o próximo nó (encadeamento)


# Hash Table com encadeamento --------------------------------------------------
class HashTableChained:
    def __init__(self, capacity=4, load_factor=0.75):
        self.capacity = capacity               # número inicial de buckets
        self.size = 0                          # quantidade atual de elementos
        self.buckets = [None] * capacity       # lista de buckets (listas encadeadas)
        self.load_factor = load_factor         # limiar de carga para rehash
        self.comparacoes = 0                   # contador de comparações para teste

    # Função hash simples baseada no hash nativo do Python -------------------------------------------
    def _hash(self, key):
        return hash(key) % self.capacity

    # Rehash automático aumentando a capacidade (dobrando) -------------------------------------------
    def _rehash(self):
        old_buckets = self.buckets            # guarda os buckets antigos
        self.capacity *= 2                     # dobra a capacidade
        self.buckets = [None] * self.capacity # cria nova tabela vazia
        self.size = 0                          # reset do tamanho

        # Reinserir todos os elementos na nova tabela
        for head in old_buckets:               # percorre cada bucket antigo
            node = head
            while node:
                self.put(node.key, node.value) # insere na nova tabela
                node = node.next

    # Inserir ou atualizar um par (key, value) --------------------------------------------------
    def put(self, key, value):
        # checar se precisa de rehash
        if self.size / self.capacity >= self.load_factor:
            self._rehash()                    # rehash se fator de carga exceder

        index = self._hash(key)
        node = self.buckets[index]
        prev = None

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1             # contar comparação
            if node.key == key:               # se encontrar atualiza o valor
                node.value = value            # atualiza valor existente
                return
            prev = node
            node = node.next

        # se não encontrar cria um novo nó no final da lista encadeada
        new_node = Node(key, value)
        if prev:
            prev.next = new_node              # adiciona ao final da lista
        else:
            self.buckets[index] = new_node    # bucket estava vazio
        self.size += 1

    # Recuperar valor pela chave --------------------------------------------------
    def get(self, key):
        index = self._hash(key)
        node = self.buckets[index]

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1
            if node.key == key:               # se a chave do nó atual = chave procurada
                return node.value             # retorna valor
            node = node.next
        return None                            # chave não encontrada

    # Remover par pela chave -----------------------------------------------------
    def delete(self, key):
        index = self._hash(key)
        node = self.buckets[index]
        prev = None

        # percorre todos os elementos procurando a chave
        while node:
            self.comparacoes += 1
            if node.key == key:               # se a chave do nó atual = chave procurada
                if prev:                      # se não é primeiro nó da lista
                    prev.next = node.next
                else:                          # se é o primeiro nó do bucket
                    self.buckets[index] = node.next
                self.size -= 1
                return True                   # removeu com sucesso
            prev = node
            node = node.next
        return False                          # chave não encontrada

    # Retorna o número de elementos na tabela --------------------------------------------------
    def __len__(self):
        return self.size
```

```python
# TESTANDO ------------------------------
# Criar tabela pequena para forçar rehash
ht = HashTableChained(capacity=2, load_factor=0.7)
comparacoes_antes = ht.comparacoes

# Inserir elementos
for i in range(10):
    ht.put(f"chave{i}", i)

# Comparações para inserção
comparacoes_insercao = ht.comparacoes
print(f"Comparações na inserção de 10 elementos: {comparacoes_insercao}")

# Comparações para busca de todas as chaves
ht.comparacoes
for i in range(10):
    ht.get(f"chave{i}")
comparacoes_busca = ht.comparacoes
print(f"Comparações na busca de 10 elementos: {comparacoes_busca}")

# Fator de carga atual
fator_carga = ht.size / ht.capacity
print(f"Fator de carga atual: {fator_carga:.2f}")

# Relação custo esperado
print("Custo esperado de busca e inserção é O(1 + α), ou seja:")
print(f"O(1 + {fator_carga:.2f}) ≈ O({1 + fator_carga:.2f}) comparações por operação")
```
