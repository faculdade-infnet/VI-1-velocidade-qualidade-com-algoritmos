# 1. Busca linear e Busca Binária

```python
import random

# BUSCA LINEAR -------------------------------------------------------------
def busca_linear(arr, target):
    n = len(arr)
    cont = 0                    # conta número de passos executados

    for i in range(n):
        cont += 1
        if arr[i] == target:
            return i, cont      # retorna posição encontrada
    return -1, cont             # retorna se não encontrou nada

# PRÉ-CONDIÇÃO: vetor deve estar ordenado ----------------------------------
# FALHA RÁPIDA: se NÃO estiver ordenado ------------------------------------
def esta_ordenado(arr):
    # percorre o comprimentodo array,
    for i in range(len(arr)-1):
        # caso algum valor nõa esteja ordenado (menor --> maior)
        if arr[i] > arr[i+1]:
            return False
    return True

# BUSCA BINÁRIA ------------------------------------------------------------
def busca_binaria(arr, target):
    # pré-condição
    if not esta_ordenado(arr):
        return -1, 0        # falha rápida

    inicio = 0
    fim = len(arr)-1
    cont = 0                # conta número de passos executados

    while inicio <= fim:
        cont += 1
        meio = (inicio + fim)//2    # divisão inteira

        if arr[meio] == target:
            return meio, cont
        if arr[meio] > target:
            fim = meio - 1
        else:
            inicio = meio + 1

    return -1, cont


# GERAR VETOR --------------------------------------------------------------
def gerar_vetor(n, ordenado=False):
    arr = [random.randint(0, n) for _ in range(n)]
    return sorted(arr) if ordenado else arr
```

```python
# TESTANDO -----------------------------------------------------------------
buscar_valor = 11        # valor a ser buscado
encontrado = False       # verifica se valor foi encontrado

print(f"Nº de comparações ao buscar ({buscar_valor})")
print(f"---------------------------------------")

for n in [100, 1000, 10000, 100000]:
    arr = gerar_vetor(n, ordenado=True) # gera vetor ordenado ou não ordenado

    # faz busca linear e retrona posição e número de comparações
    pos_lin, cont_lin = busca_linear(arr, buscar_valor)
    # faz busca binária e retrona posição e número de comparações
    pos_bin, cont_bin = busca_binaria(arr, buscar_valor)

    # verifica se encontrou o vlaor em alguma busca
    if pos_lin != -1 or pos_bin != -1:
        encontrado = True

    print(f"n={n:<6} | linear={cont_lin:<6} | binaria={cont_bin}")

# mensagem final
print("\nResultado da busca:")
if encontrado:
    print(f"Nº {buscar_valor}, encontrado.")
else:
    print(f"Nº {buscar_valor}, NÃO encontrado.")
```
