# 7. Lista duplamente encadeada + Deque

```python
# Nodo da lista duplamente encadeada --------------------------
class Nodo:
    def __init__(self, valor):
        self.valor = valor         # guarda o valor do nodo
        self.anterior = None       # ponteiro pro nodo anterior
        self.proximo = None        # ponteiro pro próximo nodo


# Lista duplamente encadeada --------------------------
class ListaDuplamenteEncadeada:
    def __init__(self):
        self.cabeca = None         # começo da lista
        self.cauda = None          # fim da lista
        self.tamanho = 0           # contador de elementos (igual na pilha TP2)

    def esta_vazia(self):
        return self.tamanho == 0   # True se não tem nada

    # inserir no começo da lista -------------------------------
    def inserir_inicio(self, valor):
        novo = Nodo(valor)
        if self.esta_vazia():       # se lista vazia, cabeça e cauda viram o novo nodo
            self.cabeca = self.cauda = novo
        else:
            novo.proximo = self.cabeca
            self.cabeca.anterior = novo
            self.cabeca = novo
        self.tamanho += 1

    # inserir no fim da lista -------------------------------
    def inserir_fim(self, valor):
        novo = Nodo(valor)
        # Se a lista estiver vazia, então ele vira a cabeça e a cauda ao mesmo tempo
        if self.esta_vazia():
            self.cabeca = self.cauda = novo
        else:       # O novo nodo aponta pro antigo último nodo (anterior)
            novo.anterior = self.cauda
            self.cauda.proximo = novo
            self.cauda = novo
        self.tamanho += 1

    # remover do começo da lista -------------------------------
    def remover_inicio(self):
        if self.esta_vazia():
            print("Lista vazia! Nada pra remover.")
            return None

        removido = self.cabeca  # Guarda o nodo que vamos remover
        if self.cabeca == self.cauda:   # só tinha um elemento
            self.cabeca = self.cauda = None
        else:
            self.cabeca = self.cabeca.proximo
            self.cabeca.anterior = None
        self.tamanho -= 1
        return removido.valor

    # remover do fim da lista -------------------------------
    def remover_fim(self):
        if self.esta_vazia():
            print("Lista vazia! Nada pra remover.")
            return None

        removido = self.cauda       # Guarda o nodo que vamos remover

        # Se são iguais é pq só tinha um elemento
        if self.cabeca == self.cauda:
            self.cabeca = self.cauda = None
        else:           # Senao, então tem mais elementos, a cabeça agora vira o próximo nodo
            self.cauda = self.cauda.anterior
            self.cauda.proximo = None

        self.tamanho -= 1       # -1, porque removemos um elemento
        return removido.valor

    # imprimir a lista -------------------------------
    def imprimir(self):
        atual = self.cabeca
        valores = []
        while atual:
            valores.append(atual.valor)
            atual = atual.proximo
        print("Lista: ", valores)


    # Verificação das invariantes estruturais  --------------------------
    def verificar_invariantes(self):
        # Se a lista estiver vazia, a cabeça e a cauda devem ser None
        if self.esta_vazia():
            if self.cabeca is not None or self.cauda is not None:
                print("Erro: lista vazia mas cabeca ou cauda não são None!")
            return True  # tudo certo, lista vazia

        # checando cabeça e cauda
        if self.cabeca.anterior is not None:
            print("Erro: cabeca.anterior não é None")  # cabeça não deve ter anterior
        if self.cauda.proximo is not None:
            print("Erro: cauda.proximo não é None")    # cauda não deve ter próximo

        atual = self.cabeca
        contador = 0
        while atual:        # percorrer a lista e checar ponteiros
            contador += 1
            # checar se proximo nodo aponta de volta pro atual
            if atual.proximo and atual.proximo.anterior != atual:
                print(f"Erro: nodo {atual.valor}.proximo.anterior != nodo atual")
            # checar se nodo anterior aponta de volta pro atual
            if atual.anterior and atual.anterior.proximo != atual:
                print(f"Erro: nodo {atual.valor}.anterior.proximo != nodo atual")
            atual = atual.proximo

        # checar se o tamanho registrado bate com a contagem real
        if contador != self.tamanho:
            print(f"Erro: tamanho armazenado ({self.tamanho}) != contagem real ({contador})")

        # retorno True se tudo bateu certinho, False se deu algum problema
        return contador == self.tamanho


# Deque usando a Lista Duplamente encadeada --------------------------
class Deque:
    def __init__(self):
        self.lista = ListaDuplamenteEncadeada()  # usa a lista dupla

    def inserir_esquerda(self, valor):      # inserir à esquerda (começo da lista)
        self.lista.inserir_inicio(valor)

    def inserir_direita(self, valor):       # inserir à direita (fim da lista)
        self.lista.inserir_fim(valor)

    def remover_esquerda(self):             # remover da esquerda (começo da lista)
        return self.lista.remover_inicio()

    def remover_direita(self):              # remover da direita (fim da lista)
        return self.lista.remover_fim()

    def espiar_esquerda(self):              # espiar o elemento da esquerda
        if self.lista.esta_vazia():
            print("Deque vazio!")
            return None
        return self.lista.cabeca.valor

    def espiar_direita(self):               # espiar o elemento da direita
        if self.lista.esta_vazia():
            print("Deque vazio!")
            return None
        return self.lista.cauda.valor

    # imprimir deque
    def imprimir(self):
        self.lista.imprimir()

    # método auxiliar para verificar invariantes
    def checar_invariantes(self):
        print("Verificando invariantes...")  # comentário estilo aluno
        ok = self.lista.verificar_invariantes()
        if ok:
            print("Tudo certo! Ponteiros consistentes e tamanho correto.")
        else:
            print("Erro nas invariantes!")  # se algo estiver errado nos ponteiros ou tamanho
```

```python
# TESTANTO ----------------------------------------------
deque = Deque()

# inserções alternadas
deque.inserir_esquerda(10)
deque.inserir_direita(20)
deque.inserir_esquerda(5)
deque.inserir_direita(25)
print("Depois das inserções:")
deque.imprimir()
deque.checar_invariantes()   # verifica ponteiros + tamanho

# espiar os dois lados
print("Espiar esquerda:", deque.espiar_esquerda())
print("Espiar direita:", deque.espiar_direita())

# remoções alternadas
print("Remover esquerda:", deque.remover_esquerda())
print("Remover direita:", deque.remover_direita())
print("Depois das remoções:")
deque.imprimir()
deque.checar_invariantes()

# mais operações pra testar invariantes
deque.inserir_direita(30)
deque.inserir_esquerda(2)
deque.inserir_direita(40)
print("Inserções extras:")
deque.imprimir()
deque.checar_invariantes()  # invariantes estruturais verificadas

# verificar tamanho final
print("Tamanho final do deque:", deque.lista.tamanho)
```
