# 1. Busca linear e Busca Binária

```python
import random

# BUSCA LINEAR -------------------------------------------------------------
def busca_linear(lista, busca):
    n = len(lista)
    cont =  0 # conta número de passos executados

    for i in range(n):
        cont += 1
        if lista[i] == busca:
            return i, cont   # retorna posição e comparações
    return -1, cont          # não encontrou


# PRÉ-CONDIÇÃO: vetor deve estar ordenado ----------------------------------
# FALHA RÁPIDA: se NÃO estiver ordenado ------------------------------------
def esta_ordenado(arr):
    # percorre o comprimento do array
    for i in range(len(arr)-1):
        # caso algum valor não esteja ordenado (menor --> maior)
        if arr[i] > arr[i+1]:
            return False
    return True


# FALHA RÁPIDA (amostragem - mais barato que checar tudo) -------------------
def esta_ordenado_amostra(arr, k=10):
    n = len(arr)
    if n <= 1:
        return True

    # verifica apenas alguns pontos do vetor (não tudo)
    for _ in range(k):
        i = random.randint(0, n-2)
        if arr[i] > arr[i+1]:
            return False
    return True


# BUSCA BINÁRIA ------------------------------------------------------------
def busca_binaria(lista, busca):
    # pré-condição (falha rápida com amostragem)
    if not esta_ordenado_amostra(lista):
        return -1, 0   # falha rápida (não faz busca)

    inicio = 0
    fim = len(lista)-1
    cont =  0 # conta número de passos executados

    while inicio <= fim:
        cont += 1
        meio = (inicio + fim)//2   # divisão inteira

        if lista[meio] == busca:
            return meio, cont
        if lista[meio] > busca:
            fim = meio - 1
        else:
            inicio = meio + 1

    return -1, cont


# GERAR VETOR --------------------------------------------------------------
def gerar_vetor(n, ordenado=False):
    arr = [random.randint(0, n) for _ in range(n)]
    return sorted(arr) if ordenado else arr
```

```python
# TESTANDO -----------------------------------------------------------------
buscar_valor = [-14, 11]        # valor a ser buscado
encontrado = False       # verifica se valor foi encontrado


for valor in buscar_valor:
    print(f"\n---------------------------------------")
    print(f"Nº de comparações ao buscar ({valor})")
    print(f"---------------------------------------")

    for n in [100, 1000, 10000, 100000, 1000000]:
        arr = gerar_vetor(n, ordenado=True) # gera vetor ordenado ou não ordenado

        # faz busca linear e retorna posição e número de comparações
        pos_lin, cont_lin = busca_linear(arr, valor)
        # faz busca binária e retorna posição e número de comparações
        pos_bin, cont_bin = busca_binaria(arr, valor)

        # verifica se encontrou o valor em alguma busca
        if pos_lin != -1 or pos_bin != -1:
            encontrado = True

        print(f"n={n:<8} | linear={cont_lin:<10} | binaria={cont_bin}")


    # mensagem final
    print("\nResultado da busca:")
    if encontrado:
        print(f"Nº {valor}, encontrado.")
    else:
        print(f"Nº {valor}, NÃO encontrado.")
```
